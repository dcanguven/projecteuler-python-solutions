## Initial approach

* avoid recursive evaluation completely
* the function follows a block pattern from the top down
* values close to b are easy because the recursive calls immediately cross the boundary
* each full block moving left increases the added offset by a fixed amount
* this turns the whole sum into a floor-sum problem
* compute the floor-sum directly with division
* take only the last nine digits at the end

```python
def S(a, b, c, mod=None):
    q, r = divmod(b, a)

    floor_sum = a * q * (q - 1) // 2 + q * (r + 1)

    total = b * (b + 1) // 2
    total += c * (b + 1)
    total += 2 * c * floor_sum

    if mod is not None:
        total %= mod

    return total

assert S(50, 2000, 40) == 5204240

a = 21 ** 7
b = 7 ** 21
c = 12 ** 7
```

```python
%%time
result = S(a, b, c, 10 ** 9)
print("Result:", result)
```
