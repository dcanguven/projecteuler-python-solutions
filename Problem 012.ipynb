## Clean and tested code

```python
%%time

import math

def find_divisors(n):
    count = 0
    sqrt_n = int(math.sqrt(n))
    for i in range(1, sqrt_n + 1):
        if n % i == 0:
            count += 2 if i != n // i else 1
    return count

def find_triangle_number(min_divisors):
    n = 1
    while True:
        triangle = n * (n + 1) // 2
        if find_divisors(triangle) > min_divisors:
            return triangle
        n += 1

result = find_triangle_number(500)
print(result)
```
