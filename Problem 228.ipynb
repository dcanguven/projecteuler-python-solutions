## Initial approach

* see each regular polygon as a solid shape centered at the same point
* when you add two such shapes, the result is still a solid shape with straight sides
* the sides of the new shape come from combining the side directions of the originals
* a regular polygon has evenly spaced side directions, and opposite sides point the same way
* when you add many regular polygons together, every new side direction that appears adds new sides
* some polygons introduce directions that were already present, others introduce completely new ones
* the task becomes counting how many genuinely new directions appear as you go through all polygons
* the total number of sides is twice the number of unique directions you end up with

```python
import math

def totients_upto(n):
    phi = list(range(n + 1))
    for i in range(2, n + 1):
        if phi[i] == i:
            for j in range(i, n + 1, i):
                phi[j] -= phi[j] // i
    return phi

def sides_of_sum(m):
    if m <= 1:
        return 0
    phi = totients_upto(m)
    total = 1
    for n in range(2, m + 1):
        total += phi[n]
    return 2 * total
```

```python
%%time
result = sides_of_sum(1000)
print("Result:", result)
```
