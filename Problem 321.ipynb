## Initial approach

* the minimum move count for size n has a simple closed pattern
* we need the values of n where that move count is triangular
* after rewriting the triangular condition, it becomes a Pell-type equation
* Pell solutions grow very fast, so only a few generations are needed
* there are two interleaving solution branches
* generate both branches and sort the resulting n values
* take the first 40 values and sum them

```python
def generate_terms(count):
    terms = []
    seeds = [(5, 2), (11, 4)]

    for y, x in seeds:
        for _ in range(count):
            terms.append(x - 1)
            y, x = 3 * y + 8 * x, y + 3 * x

    terms = sorted(set(terms))
    return terms[:count]

def solve():
    terms = generate_terms(40)
    return sum(terms)

first_five = generate_terms(5)
assert first_five == [1, 3, 10, 22, 63]
assert sum(first_five) == 99
```

```python
%%time
result = solve()
print("Result:", result)
```
