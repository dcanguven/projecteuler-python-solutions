## Initial approach

* start with a number n
* multiply it by 1,2,3... and joint the results
* stop when joined number reaches 9 digits
* check if it uses all digits 1 to 9, only once.
* save the biggest such number

```python
def is_pandigital(s):
    return ''.join(sorted(s)) == '123456789'

max_pandigital = 0

for n in range(1, 10000):
    concat = ''
    i = 1
    while len(concat) < 9:
        concat += str(n * i)
        i += 1
    if len(concat) == 9 and is_pandigital(concat):
        max_pandigital = max(max_pandigital, int(concat))

print("Result:", max_pandigital)
```
