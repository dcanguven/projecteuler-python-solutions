## Initial approach

* define bouncy number first with a function
* so define increasing and decreasing number first

```python
def is_increasing(num):
    for i in range(len(str(num))-1):
        if str(num)[i]<=str(num)[i+1]:
            continue
        else:
            return False
    return True

def is_decreasing(num):
    for i in range(len(str(num))-1):
        if str(num)[i]>=str(num)[i+1]:
            continue
        else:
            return False
    return True
```

* Try:

```python
print("is True?", is_increasing(12345))
print("is False?", is_decreasing(12345))
print("is False?", is_increasing(54321))
print("is True?", is_decreasing(54321))
```

* So bouncy is:

```python
def is_bouncy(num):
    if not is_increasing(num) and not is_decreasing(num):
        return True
    return False
```

* check the 50% case

```python
total = 0
for i in range(100, 10000):
    if is_bouncy(i):
        total += 1
    if total / i == 0.5:
        print("First time ratio 50% at:", i)
        break
```

* check the 90% case

```python
total = 0
for i in range(100, 100000):
    if is_bouncy(i):
        total += 1
    if total / i == 0.9:
        print("First time ratio 90% at:", i)
        break
```

* check the result, 99% case:

```python
%%time

total = 0
for i in range(100, 10**10):
    if is_bouncy(i):
        total += 1
    if total / i == 0.99:
        print("First time ratio 99% at:", i)
        break
```
