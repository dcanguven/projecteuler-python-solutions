## Initial approach

* iterate through all numbers between 10 and 99 for both a and b
* when a and b has the same number:
    * divide them
    * remove the same number
    * divide them
    * compare divisions

```python
%%time

from math import gcd

num_product = 1
den_product = 1

for a in range(10, 100):
    for b in range(a + 1, 100):
        a_str = str(a)
        b_str = str(b)

        common_digits = set(a_str) & set(b_str) - {'0'}

        if common_digits:
            for digit in common_digits:
                a_new = a_str.replace(digit, '', 1)
                b_new = b_str.replace(digit, '', 1)

                if a_new and b_new and a_new.isdigit() and b_new.isdigit():
                    a2 = int(a_new)
                    b2 = int(b_new)

                    if b2 != 0 and a * b2 == a2 * b:
                        num_product *= a
                        den_product *= b

common_divisor = gcd(num_product, den_product)
simplified_denominator = den_product // common_divisor

print(simplified_denominator)
```
