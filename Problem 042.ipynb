## Initial approach

* Understand the example in the question.

```python
def n_term(n):
    return int(n*0.5*(n+1))
```

```python
for i in range(1,11):
    print(n_term(i))
```

* Looks good. Now put the alphabet into a dictionary.  
* Use chr() for iteration

```python
print(chr(65))
print(chr(65+25))
```

```python
alphabet = dict()
order = 1
for i in range(65, 65+26):
    alphabet[chr(i)]=order
    order+=1

alphabet
```

* Read and convert the text into list

```python
with open("0042_words.txt", "r") as file:
    content=file.read()
    words = content.replace('"', '').split(',')
```

```python
print(len(words))
```

* Find the maximum word length

```python
max_length_word = 0
for i in words:
    if len(i)>max_length_word:
        max_length_word=len(i)

max_length_word
```

* The longest word has 14 letters, and assuming all are Z
* The maximum value would be 14  26 = 364.
* This is the upper limit for iteration.
* Lets create a list of triangle numbers up to at least the 28th term.
* 28th term exceeds 364, the longest possible word's value.

```python
list_of_terms = [n_term(x) for x in range(1,28)]
```

* Now iterate through words list

```python
%%time

result = 0
for word in words:
    total = 0
    while len(word)>0:
        total+=alphabet[word[0]]
        word=word[1:]
    if total in list_of_terms:
        result+=1

print("Result:", result)
```
