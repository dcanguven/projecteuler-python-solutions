## Initial approach

* turn the three consecutive ap terms into td, t, t+d and get n = 4td  t^2 with t the middle term and d the common difference
* set u = t and v = 4d  t so n becomes uv
* d is integer only if u + v is divisible by 4
* positivity and consecutive terms give the extra condition v < 3u
* each ordered factor pair (u, v) of n that satisfies (u + v) mod 4 = 0 and v < 3u gives one solution
* for every n below one million count such pairs and keep those with exactly ten solutions
* sum how many n meet this count

```python
import math

def solutions_count(n):
    c = 0
    r = int(math.isqrt(n))
    for a in range(1, r + 1):
        if n % a == 0:
            b = n // a
            if ((a + b) & 3) == 0 and 3 * a > b:
                c += 1
            if a != b and ((a + b) & 3) == 0 and 3 * b > a:
                c += 1
    return c
```

```python
%%time
limit = 10**6
result = 0
for n in range(1, limit):
    if solutions_count(n) == 10:
        result += 1

print("Result:", result)
```
