## Initial approach

* we turn the product into a sum using logs, which makes the optimisation easier
* we apply a constraint method that tells us each variable must scale in a simple pattern
* the result shows every value in the optimal tuple is proportional to its index
* using the sum constraint fixes the proportionality factor
* substituting these optimal values back into the product gives a closed form
* this expression can be evaluated exactly for each case we need
* we then take the integer part for every case and add them up

```python
def floor_P_m(m):
    s = m * (m + 1) // 2
    num = pow(2, s)
    for k in range(1, m + 1):
        num *= k ** k
    den = pow(m + 1, s)
    return num // den

total = 0
for m in range(2, 16):
    total += floor_P_m(m)
```

```python
%%time
result = total
print("Result:", result)
```
