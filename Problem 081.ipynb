## Initial approach

Numpy is faster for this task, here is why:

*  numpy is written in c language, so it is faster than python lists.  
*  it can do math on arrays without using loops.  
*  it uses less memory than normal python lists.  

```python
import numpy as np

matrix = np.loadtxt("0081_matrix.txt", delimiter=",", dtype=int)

n = matrix.shape[0]

for i in range(n - 2, -1, -1):
    matrix[i, -1] += matrix[i + 1, -1]
    matrix[-1, i] += matrix[-1, i + 1]

for i in range(n - 2, -1, -1):
    for j in range(n - 2, -1, -1):
        matrix[i, j] += min(matrix[i + 1, j], matrix[i, j + 1])

print(matrix[0, 0])
```
