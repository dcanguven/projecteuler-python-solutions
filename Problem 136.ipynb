## Initial approach

* write x y z as t−d t t+d so n becomes 4td − t^2
* set u = t and v = 4d − t to turn it into n = u·v
* d is integer only if u + v is divisible by 4
* positivity of the smallest term needs t − d > 0 which gives v < 3u
* each ordered factor pair (u v) with (u + v) mod 4 = 0 and v < 3u gives one solution
* count these pairs for each n and keep those with exactly one such pair
* search all n below fifty million and tally how many have a unique solution

```python
%%time

import sys

N = 50_000_000
cnt = bytearray(N + 1)

for u in range(1, N + 1):
    vmax = N // u
    t = 3 * u - 1
    if vmax > t:
        vmax = t
    if vmax <= 0:
        break
    r = (-u) & 3
    v0 = r if r != 0 else 4
    if v0 > vmax:
        continue
    for v in range(v0, vmax + 1, 4):
        n = u * v
        cnt[n] += 1

result = sum(1 for x in cnt[1:] if x == 1)
print("Result:", result)
```
