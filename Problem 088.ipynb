## Initial approach

* define search limit at least double of max k value, 12000.
* need a function that tries all possible number combinations
* then we need to start trying factors from 2
* factor 1 results with infinite loops

```python
def find_min_product_sum(k_max):
    limit = 2 * k_max
    min_product = [float('inf')] * (k_max + 1)

    def generate(product, total, count, start):
        k = product - total + count
        if k <= k_max:
            if product < min_product[k]:
                min_product[k] = product
            for i in range(start, limit // product + 1):
                generate(product * i, total + i, count + 1, i)

    generate(1, 0, 0, 2)

    unique_values = set(min_product[2:])
    return sum(unique_values)

print(find_min_product_sum(12000))
```
