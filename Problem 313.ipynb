## Initial approach

* avoid simulating the sliding board directly
* use the observed minimum-move pattern for rectangular grids
* for each prime, count how many grid sizes produce that prime square
* the count grows by a simple value based on the prime
* handle the small prime 3 as a special case
* use a sieve to generate all primes below one million
* sum the grid counts for all valid primes

```python
def prime_sieve(limit):
    is_prime = [True] * limit
    is_prime[0] = False
    is_prime[1] = False

    p = 2
    while p * p < limit:
        if is_prime[p]:
            for multiple in range(p * p, limit, p):
                is_prime[multiple] = False
        p += 1

    return [i for i in range(limit) if is_prime[i]]

def solve(limit):
    primes = prime_sieve(limit)
    total = 0

    for p in primes:
        if p == 3:
            total += 2
        else:
            total += (p * p - 1) // 12

    return total
```

```python
%%time
result = solve(10 ** 6)
print("Result:", result)
```
