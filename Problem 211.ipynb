## Initial approach

* treat s(n) as sigma2(n), the sum of squares of all divisors of n
* create a python list sigma2 of length 64,000,001 filled with zeros
* run a sieve style loop, for each d from 1 to limit add dÂ² to every multiple of d in the list
* this avoids prime factoring each n, you reuse the same work across many numbers
* after filling the list, check each sigma2[n] with math.isqrt and verify root*root == sigma2[n]
* whenever it is a perfect square, add n to the running total and print the total at the end

```python
%%time

import math

limit = 64_000_000
sigma2 = [0] * (limit + 1)

for d in range(1, limit + 1):
    dsq = d * d
    for m in range(d, limit + 1, d):
        sigma2[m] += dsq

total = 0
for n in range(1, limit + 1):
    v = sigma2[n]
    r = math.isqrt(v)
    if r * r == v:
        total += n

print("Result:", total)
````

* it takes too long
* logic will be optimized.
