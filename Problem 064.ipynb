## Initial approach

* since we pass rational square roots, eliminate them from the iteration
* iterate from 1 to 10.000
* take the square root each time and separate the integer part
* for remains, find the length of periods.
* if period length is odd, count them.

```python
import math

def get_period_length(n):
    m = 0
    d = 1
    a0 = int(math.sqrt(n))
    a = a0
    period = 0

    if a0 * a0 == n:
        return 0

    while True:
        m = d * a - m
        d = (n - m * m) // d
        a = (a0 + m) // d
        period += 1
        if a == 2 * a0:
            break

    return period
```

* a0 - The integer part of sqrt(n), calculated as int(math.sqrt(n)).
* m - The remainder part, starting with 0 and updated in each step.
* d - The denominator, starting with 1 and updated in each step.
* a - The integer part in each step, calculated using m, d, and a.
* The period ends when a becomes equal to 2  a0.

```python
odd_period_count = 0

for n in range(2, 10001):
    if int(math.sqrt(n)) ** 2 == n:
        continue
    period_length = get_period_length(n)
    if period_length % 2 == 1:
        odd_period_count += 1

print(odd_period_count)
```
