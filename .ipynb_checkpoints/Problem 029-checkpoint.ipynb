## Initial approach

* iterate "a", over all integers
* for each a, iterate "b" all over integers
* compute a^b
* handle duplicates
* return length for distict terms

```python
results = set()

for a in range(2, 101):
    for b in range(2, 101):
        value = a ** b
        results.add(value)

print(len(results))
```
