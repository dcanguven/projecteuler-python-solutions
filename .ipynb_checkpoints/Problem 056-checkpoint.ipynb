## Initial approach

* numbers looks too large to compute, but lets try to iterate

```python
%%time

max_digit_sum = 0

for a in range(1,100):
    for b in range(1,100):
        num = a**b        
        sum_of_digit = 0 
        for digit in str(num):
            sum_of_digit += int(digit)
        if sum_of_digit > max_digit_sum:
            max_digit_sum = sum_of_digit

max_digit_sum
```

that was fast. even the numbers are too big.
