## What is a pandigital number?

Lets remember problem 38.

   

So specifically:
* It must be max 9 digits long.
* It must use each digit from 1 to 9 once and only once.
* It must not include 0.

Additionally, it must be prime.

Start with is_prime function.

```python
def is_prime(n):
    if n < 2:
        return False
    if n == 2 or n == 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    for i in range(5, int(n**0.5) + 1, 6):
        if n % i == 0 or n % (i + 2) == 0:
            return False
    return True
```

Let's think about the iteration logic.

* 9-digit like 987654321 is divisible by 3.
* Same for 8, 6, and 5-digit pandigital numbers.
* 7-digit is the best candidate.
* Write a function for 7-digit pandigital number check.
* Start iteration from 7654321.

```python
def is_pandigital(n):
    s = str(n)
    return set(s) == set('1234567') and len(s) == 7
```

```python
for num in range(7654321, 1234566, -2):
    if is_pandigital(num) and is_prime(num):
        print(num)
        break
```
