* since all numbers from 1 to 9 must be included, then we need to check only 9! permutations
* options:
    * 2 digits x 3 digits = 4 digits
    * 1 digit x 4 digits = 4 digits
* set is useful to not to double-count products

````python
%%time

from itertools import permutations

products = set()

digits = "123456789"

for i in permutations(digits):
    s = "".join(i)

    # for 2 digits x 3 digits = 4 digits


    a = int(s[0:2])
    b = int(s[2:5])
    c = int(s[5:])
    if a*b == c:
        products.add(c)

    # for 1 digit x 4 digits = 4 digits
    a = int(s[0])
    b = int(s[1:5])
    c = int(s[5:])
    if a * b == c:
        products.add(c)

````

````python
sum(products)
````
