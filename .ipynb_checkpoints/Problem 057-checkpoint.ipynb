## Initial approach

Lets try to write the formula

Xn+1 = Xn + 2Kn

Kn+1 = Xn + Kn

Put this into a function

```python
def xn_plus_one(xn, kn):
    return xn + 2*kn

def kn_plus_one(xn, kn):
    return xn + kn

xn = 1
kn = 1
print(xn_plus_one(xn,kn))
print(kn_plus_one(xn,kn))
```

Now iterate for the solution.

```python
xn = 1
kn = 1
count = 0

for _ in range(1000):
    xn, kn = xn_plus_one(xn, kn), kn_plus_one(xn, kn)
    if len(str(xn)) > len(str(kn)):
        count += 1

print(count)
```
