* For a large number like 600851475143, it's checking every single number up to n-1 to see if it's a factor.
* This is extremely slow and unnecessary, especially when most numbers are not factors!
* Code above needs to be optimized.

## Clean and tested code

```python
%%time

import math

def largest_prime_factor(n):
    while n % 2 == 0:
        max_prime = 2
        n //= 2
    factor = 3
    max_factor = math.isqrt(n) + 1
    while factor <= max_factor and n != 1:
        if n % factor == 0:
            max_prime = factor
            while n % factor == 0:
                n //= factor
            max_factor = math.isqrt(n) + 1
        factor += 2
    if n > 1:
        max_prime = n
    return max_prime

print(largest_prime_factor(600851475143))
```
