## Initial approach

* we only care about sums modulo 250, so reduce every i^i using pow(i, i, 250)
* keep a dp list of size 250 where dp[r] counts subsets giving remainder r
* start with dp[0] = 1 for the empty subset
* for each i from 1 to 250250, update dp by adding the contribution of remainder r
* always take modulo 10^16 to keep only the last 16 digits
* at the end dp[0] counts all subsets divisible by 250, subtract 1 to remove the empty subset
* print the result

```python
MOD = 10**16
N = 250250
M = 250

def solve():
    dp = [0] * M
    dp[0] = 1

    for i in range(1, N + 1):
        r = pow(i, i, M)
        new_dp = dp[:]
        for s in range(M):
            new_dp[(s + r) % M] = (new_dp[(s + r) % M] + dp[s]) % MOD
        dp = new_dp

    result = (dp[0] - 1) % MOD
    print("Result:", result)

solve()
```
