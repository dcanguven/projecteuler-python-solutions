## Initial approach

* Generate two functions:
    * checks if it is palindrome in decimal
    * checks if it is palindrome in binary
* convert input into integer to eliminate zeros in binary form
* compare two, and add decimal into total sum
* iterate through 1M.

```python
def check_if_palindrome_in_decimal(s):
    
    # i've got this from my leetcode algorithm notes
    s=str(s)
    left = 0
    right = len(s) - 1

    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    
    return True
```

```python
def check_if_palindrome_in_binary(s):
    
    # i've got this from my leetcode algorithm notes
    s=int(bin(s)[2:])
    s=str(s)
    left = 0
    right = len(s) - 1

    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    
    return True
```

```python
print(check_if_palindrome_in_decimal(585))
print(check_if_palindrome_in_binary(585))
```

Now, iterate.

```python
%%time

result = 0

for i in range(1,1_000_000):
    if check_if_palindrome_in_decimal(i)==check_if_palindrome_in_binary(i)==True:
        result+=i

result
```
