## Initial approach

- `a` = 200p coin  
- `b` = 100p coin  
- `c` = 50p coin  
- `d` = 20p coin  
- `e` = 10p coin  
- `f` = 5p coin  
- `g` = 2p coin  
- `h` = 1p coin

* Each loop adds one type of coin from the list.
* Make sure we don't go over the total of 200.
* If total is exactly 200, increase count by 1.

```python
%%time

count = 0

for a in range(0, 200 + 1, 200):
    for b in range(0, 200 - a + 1, 100):
        for c in range(0, 200 - a - b + 1, 50):
            for d in range(0, 200 - a - b - c + 1, 20):
                for e in range(0, 200 - a - b - c - d + 1, 10):
                    for f in range(0, 200 - a - b - c - d - e + 1, 5):
                        for g in range(0, 200 - a - b - c - d - e - f + 1, 2):
                            for h in range(0, 200 - a - b - c - d - e - f - g + 1, 1):
                                if a + b + c + d + e + f + g + h == 200:
                                    count += 1

print(count)
```
