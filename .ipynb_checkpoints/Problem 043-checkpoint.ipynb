## Initial approach

Lets remember pandigital numbers from Problem 38:  

So specifically:

It must be max 9 digits long.  
It must use each digit from 1 to 9 once and only once.

* All permutations should be generated
* Convert each one into string
* Check divisibility rules one by one
* If conditions are met, add to total

```python
%%time

from itertools import permutations

total = 0
for p in permutations('0123456789'):
    number = ''.join(p)
    if (int(number[1:4]) % 2 == 0 and
        int(number[2:5]) % 3 == 0 and
        int(number[3:6]) % 5 == 0 and
        int(number[4:7]) % 7 == 0 and
        int(number[5:8]) % 11 == 0 and
        int(number[6:9]) % 13 == 0 and
        int(number[7:10]) % 17 == 0):
        total += int(number)

print(total)
```
