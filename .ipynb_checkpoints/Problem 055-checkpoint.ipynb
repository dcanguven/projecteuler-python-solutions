## Initial approach

* Problem definition: how many numbers below 10,000 become a palindrome through the reverse-and-add process in less than 50 steps
* define is_palindrome and reverse_number functions
* check number + reverse(number) is palindrome
* if not, check 50 times
* if not in 50 iteration, it's a lychrel number candidate
* iterate from 1 to 9999 and count candidates

```python
def is_palindrome(n):
    s = str(n)
    return s == s[::-1]

def reverse_number(n):
    return int(str(n)[::-1])
```

```python
lychrel_count = 0

for i in range(1, 10000):
    n = i
    for _ in range(50):
        n = n + reverse_number(n)
        if is_palindrome(n):
            break
    else:
        lychrel_count += 1

print("Answer is:", lychrel_count)
```
