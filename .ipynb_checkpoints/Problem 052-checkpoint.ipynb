## Initial approach

x, 2x, 3x, 4x, 5x, 6x means 6 different number combinations.    
that also means starting point has 6 digits

```python
%%time

for i in range(100000, 1000000):
    if sorted(str(i))==sorted(str(2*i))==sorted(str(3*i))==sorted(str(4*i))==sorted(str(5*i))==sorted(str(6*i)):
        print(i)
```

Easy :)
