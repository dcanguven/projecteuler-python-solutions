## Initial approach

```python
def fibonacci(limit):
    k0 = 0
    k1 = 1
    num = 1
    list1 = [0]
    while num<limit:
        list1.append(num)
        num = k0 + k1
        k0 = k1
        k1 = num
    return list1

total = 0
for i in fibonacci(4_000_000):
    if i%2 == 0:
        total+=i

print(total)
```

## Clean and tested code

```python
def even_fibonacci_sum(limit):
    a, b = 0, 1
    total = 0
    while b < limit:
        if b % 2 == 0:
            total += b
        a, b = b, a + b
    return total

print(even_fibonacci_sum(4_000_000))
```
