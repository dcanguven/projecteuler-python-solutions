## Initial approach

Function for checking primes.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
```

List all primes lower than 1 millon.

```python
list_of_primes = [x for x in range(2,1_000_000) if is_prime(x)]
len(list_of_primes)
```

Define the result list.

Find all sums up to 1 million.    
Try one by one.

```python
%%time

max_length = 0
max_prime = 0

for i in range(len(list_of_primes)):
    total = 0
    for j in range(i, len(list_of_primes)):
        total += list_of_primes[j]
        if total > 1_000_000:
            break
        if is_prime(total):
            length = j - i + 1
            if length > max_length:
                max_length = length
                max_prime = total

print("Result is:", max_prime)
```
