## Initial approach

* check the example

```python
from math import factorial
factorial_sum = 0
num = 145
for i in str(num):
    factorial_sum += factorial(int(i))

if factorial_sum == num:
    print(True) 
```

* build a function

```python
def check_num(num):
    factorial_sum = 0
    for i in str(num):
        factorial_sum += factorial(int(i))
    if factorial_sum == num:
        return num

# check the example:
check_num(145)
```

* how to select limits of iteration?
* 9! is maximum number to product with digit number.

```python
import math
math.factorial(9)*8
```

* 8 x 9! is less than minimum 8 digit number:

8 x 9! = 2.903.040
min 8 digit number = 10.000.000  

* so limit is 7 x 9!

```python
7 * math.factorial(9)
```

* so this is the upper limit
* 1 and 2 is excluded by problem definition
* lets build the iteration

```python
%%time

sum_of_all = 0
for i in range(3, 2540160):
    if check_num(i):
        sum_of_all+=i

print(sum_of_all)
```
