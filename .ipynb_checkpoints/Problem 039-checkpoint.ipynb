## Initial approach

* Lets understand the example.
* We are working with 3 variables; a, b, c
* p = a + b + c = 120 in the example
* 1000 is the limit for p
* a2 + b2 = c2 this condition must be applicable

```python
for a in range(1,1000):
    for b in range(1,1000):
        c = (a**2 + b**2)**0.5
        if a+b+c==120 and c==int(c):
            print(a,b,int(c))
```

we need to add an iterable to this sum:   
a+b+c==120   
and count solutions

```python
%%time

max_solutions = 0
result = 0

for p in range(1, 1001):
    count = 0
    for a in range(1, p):
        for b in range(a, p - a):
            c = p - a - b
            if a * a + b * b == c * c:
                count += 1
    if count > max_solutions:
        max_solutions = count
        result = p

print(result)
```

Correct answer is 840. Now check solutions for 840.

```python
counter = 0
for a in range(1,1000):
    for b in range(1,1000):
        c = (a**2 + b**2)**0.5
        if a+b+c==840 and c==int(c):
            print(a,b,int(c))
            counter+=1

print("Correct answer is: 840.")
print(f"There are {counter} of solutions for 840.")
```
