## Initial approach

```python
def is_prime(n):
    for i in range(2,n//2+1):
        if n%i==0:
            return False
    return True
```

```python
primes_total = 0
for i in range(2,10):
    if is_prime(i):
        print(i)
        primes_total+=i

print(primes_total)
```

```python
%%time
primes_total = 0
for i in range(2,100_000): # 2.000.000 takes a lot of time, code needs to be optimized
    if is_prime(i):
        primes_total+=i

print(primes_total)
```

## Clean and tested code

```python
%%time

def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5)+1, 2):
        if n % i == 0:
            return False
    return True

primes_total = 0
for i in range(2, 2_000_000):
    if is_prime(i):
        primes_total += i

print(primes_total)
```
