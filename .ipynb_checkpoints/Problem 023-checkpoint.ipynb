## Initial approach

#### Calculate the sum of proper divisors of a number

```python
def get_proper_divisor_sum(n):
    if n < 2:
        return 0
    divisors = set()
    divisors.add(1)
    sqrt_n = int(math.sqrt(n))
    for i in range(2, sqrt_n + 1):
        if n % i == 0:
            divisors.add(i)
            pair = n // i
            if pair != i:
                divisors.add(pair)
    return sum(divisors)
```

#### Identify all abundant numbers up to 28123

```python
import math

LIMIT = 28123
abundant_numbers = []
for i in range(12, LIMIT + 1):
    if get_proper_divisor_sum(i) > i:
        abundant_numbers.append(i)
```

#### Store all numbers: sum of two abundant numbers

```python
abundant_sums = set()
abundant_count = len(abundant_numbers)
for i in range(abundant_count):
    for j in range(i, abundant_count):
        total = abundant_numbers[i] + abundant_numbers[j]
        if total <= LIMIT:
            abundant_sums.add(total)
        else:
            break
```

#### Find the total of all positive numbers up to 28,123 that cannot be written as the sum of two abundant numbers

```python
non_abundant_sum = 0
for i in range(1, LIMIT + 1):
    if i not in abundant_sums:
        non_abundant_sum += i

non_abundant_sum
```

## Initial approach is clean, simple, and efficient enough for the given input size.
