## Initial approach

* Determine each digit step by step using factorial
* target = 999999 because it starts with 0
* target // factorial determines which digit comes next

```python
import math

digits = list("0123456789")
target = 1000000 - 1
result = ""

for i in range(10):
    factorial = math.factorial(len(digits) - 1)
    index = target // factorial
    result += digits.pop(index)
    target %= factorial
```

```python
result
```

## Initial approach is clean, simple, and efficient enough for the given input size.
