## Initial approach

* let f(n) be number of ways to tile length n
* recurrence: f(n) = f(n-1)+f(n-2)+f(n-3)+f(n-4)
* base is f(0)=1, f(n<0)=0
* lets iterate upto n=50 and return f(50)

```python
def count_ways(n):
    dp = [0] * (n + 1)
    dp[0] = 1
    for i in range(1, n + 1):
        total = 0
        if i >= 1:
            total += dp[i - 1]
        if i >= 2:
            total += dp[i - 2]
        if i >= 3:
            total += dp[i - 3]
        if i >= 4:
            total += dp[i - 4]
        dp[i] = total
    return dp[n]

n = 50
result = count_ways(n)
print("Result:", result)
```
