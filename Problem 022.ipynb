## Initial approach

```python
%%time

with open('0022_names.txt', 'r') as file:
    names = sorted(file.read().replace('"','').split(","))

alphabet_dict = {chr(i + 65): i + 1 for i in range(26)}

list_total_name_values = list()
for i in range(len(names)):
    total = 0
    for k in names[i]:
        total += alphabet_dict[k]
    list_total_name_values.append(total)

result = [x * y for x, y in zip(list_total_name_values, range(1,len(names)+1))]
sum(result)
```

```python
%%time

with open('0022_names.txt', 'r') as file:
    names = sorted(file.read().replace('"', '').split(","))

alphabet_dict = {chr(i + 65): i + 1 for i in range(26)}

name_scores = [
    sum(alphabet_dict[char] for char in name) * (idx + 1)
    for idx, name in enumerate(names)
]

print(sum(name_scores))
```

## Both approaches is clean, simple, and efficient enough for the given input size.
