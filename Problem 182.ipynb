## Initial approach

* we use p = 1009 and q = 3643, so n = pq and  = (p  1)(q  1)
* a message m is unconcealed when m^e  m (mod n), which means this must hold both mod p and mod q
* for a prime r, the equation m^e  m (mod r) has 1 + gcd(e  1, r  1) solutions for m in 0,,r1
* so for n = pq, the number of unconcealed messages is (1 + gcd(e  1, p  1))  (1 + gcd(e  1, q  1))
* valid e must satisfy 1 < e <  and gcd(e, ) = 1
* we loop over all such e, compute the number of unconcealed messages, and track the minimum value
* every e that gives this minimum is added into a running sum
* in the end we print the sum of all e that give the smallest possible number of unconcealed messages

```python
import math

p = 1009
q = 3643
phi = (p - 1) * (q - 1)

min_unconcealed = None
sum_e = 0

for e in range(2, phi):
    if math.gcd(e, phi) != 1:
        continue
    c_p = 1 + math.gcd(e - 1, p - 1)
    c_q = 1 + math.gcd(e - 1, q - 1)
    unconcealed = c_p * c_q
    if min_unconcealed is None or unconcealed < min_unconcealed:
        min_unconcealed = unconcealed
        sum_e = e
    elif unconcealed == min_unconcealed:
        sum_e += e

result = sum_e
print("Result:", result)
```
