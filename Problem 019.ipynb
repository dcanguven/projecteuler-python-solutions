## Initial approach

```python
year = 1900

months = {
    "january": 31,
    "february": 28,
    "march": 31,
    "april": 30,
    "may": 31,
    "june": 30,
    "july": 31,
    "august": 31,
    "september": 30,
    "october": 31,
    "november": 30,
    "december": 31
}

months_leap_year = {
    "january": 31,
    "february": 29,
    "march": 31,
    "april": 30,
    "may": 31,
    "june": 30,
    "july": 31,
    "august": 31,
    "september": 30,
    "october": 31,
    "november": 30,
    "december": 31
}
```

* Start from 1 Jan 1900, Monday
* Track the day of the week
* Count how many times first day of a month is Sunday

```python
def is_leap_year(y):
    return (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0)
```

```python
%%time

day = 0
count = 0

for y in range(1900, 2001):
    month_set = months_leap_year if is_leap_year(y) else months
    for m in month_set:
        if y >= 1901 and day % 7 == 6:
            count += 1
        day += month_set[m]
```
