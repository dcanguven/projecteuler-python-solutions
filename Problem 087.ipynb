Make lists from:

* squares
* cubes
* fourth powers

of primes. First define prime:

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
squares = [x**2 for x in range(2, 7071) if is_prime(x)]
cubes = [x**3 for x in range(2, 368) if is_prime(x)]
fourth_power = [x**4 for x in range(2, 84) if is_prime(x)]
```

* Now use 3 nested loops for each elements in each lists
* Check if sum < 50 million
* use set() to avoid duplicates
* return length of set

```python
%%time

limit = 50_000_000
results = set()

for a in fourth_power:
    for b in cubes:
        if a + b >= limit:
            continue
        for c in squares:
            total = a + b + c
            if total < limit:
                results.add(total)

print("Result:", len(results))
```
