## Initial approach

* arrange numbers 1 to 10 in a ring
* each line has 3 numbers
* sum of numbers in each line must be the same
* find max 16 digit string formed by all number on the lines in order.

```python
%%time

from itertools import permutations

max_string = ''

for p in permutations(range(1, 11)):
    outer = p[:5]
    inner = p[5:]
    total = outer[0] + inner[0] + inner[1]
    if all(outer[i] + inner[i] + inner[(i + 1) % 5] == total for i in range(1, 5)):
        lines = [[outer[i], inner[i], inner[(i + 1) % 5]] for i in range(5)]
        min_index = min(range(5), key=lambda i: lines[i][0])
        ordered = lines[min_index:] + lines[:min_index]
        string = ''.join(''.join(str(n) for n in line) for line in ordered)
        if len(string) == 16 and string > max_string:
            max_string = string

print("Result:", max_string)
```
