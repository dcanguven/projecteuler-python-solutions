## Initial approach

Similarities between Problem 69 and Problem 70:

- Both problems use Eulers totient function
- Both problems work with numbers up to 1,000,000
- Both problems need to compare a value involving n and totient function
- Both problems are solved faster by using prime factors
- Both problems are about finding a special number n with a maximum/minimum ratio

Key differences:

- Problem 69: Maximize n / totient function
- Problem 70: Minimize n / totient function with the extra condition that n and n / totient function are permutations of each other.

Functions from Problem 69:

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def totient(n):
    count = 0
    for i in range(1, n + 1):
        if gcd(i, n) == 1:
            count += 1
    return count
```

```python
def is_permutation(a, b):
    return sorted(str(a)) == sorted(str(b))
```

```python
%%time

min_ratio = float('inf')
result_n = 0

for n in range(2, 10000):
    phi = totient(n)
    if is_permutation(n, phi):
        ratio = n / phi
        if ratio < min_ratio:
            min_ratio = ratio
            result_n = n

print("Result:", result_n)
```
