## Initial approach

* list all primes below 150 that are 1 mod 4, because only those can appear in these squarefree numbers
* write each such prime as a sum of two squares in exactly one basic way with positive integers
* use the product identity for sums of two squares to build representations for larger numbers from smaller ones
* keep each pair in canonical form as smaller value first, so the same solution is not counted twice
* for each subset of the valid primes, generate all solution pairs for that product and add up the smaller value from each pair
* a depth first search works well here because there are only 16 eligible primes, so the subset count is manageable
* the number of solution pairs doubles each time you include another prime, so this is still fast enough in plain python
