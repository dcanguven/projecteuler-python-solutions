## Initial approach

* n : odd composite number
* p: prime number
* k: any positive number

proposal: n = p + 2 * k

Start with is_prime function

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
```

Generate composite numbers

```python
n = 9
while True:
    if not is_prime(n):
        # checking will be here
        break
    n += 2
```

Now proceed with checking.   
Start from 9 (the first odd number for which is_prime returns false)

```python
n = 9
while True:
    if not is_prime(n):
        found = False
        for p in range(2, n):
            if is_prime(p):
                k_squared = (n - p) / 2
                if k_squared == int(k_squared):
                    k = int(k_squared**0.5)
                    if k * k == k_squared:
                        found = True
                        break
        if not found:
            print("Result:",n)
            break
    n += 2
```
