## Initial approach

* set up the relation for an isosceles triangle with height h and base b using a² = h² + (b/2)²
* the problem gives h = b ± 1, so plug that into the equation
* this gives a pell-type equation: a² = (b ± 1)² + (b/2)² → 5b² ± 4b + 4 = 4a²
* simplify to form 5b² ± 4b + 4 − 4a² = 0 and rearrange into a standard Pell equation
* solve recursively by generating integer pairs (a, b) using Pell relation patterns
* collect the first 12 valid b values giving integer side lengths
* sum those b values for the result

```python
%%time

triangles = []
x1, y1 = 17, 7
x, y = x1, y1
while len(triangles) < 12:
    b = (x - 1) // 2
    triangles.append(b)
    x, y = 9*x + 20*y, 4*x + 9*y

result = sum(triangles)
print("Result:", result)
```
