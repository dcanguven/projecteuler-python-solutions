## Initial approach

* we use the fact that a triangle with one 60 degree angle satisfies a^2 + b^2  a b = c^2 for integer sides
* the inradius r of such a triangle can be expressed in terms of a, b, c, which lets us convert the condition r  n into an inequality on a and b
* primitive 60 degree triangles can be generated from two positive coprime integers (u, v) with a known parametrisation, similar to how primitive pythagorean triples are produced
* every non primitive triangle is obtained by scaling a primitive one, so for each primitive we count all allowed multiples k that keep r_k  n
* summing over all valid primitive seeds (u, v) gives T(n) efficiently
* because n is large (1053779), we rely on number theoretic bounds to limit u and v, then loop and count all admissible scalings
