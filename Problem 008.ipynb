```python
%%time

num = str(num)
greatest_product = 0

for i in range(len(num)):
    product = 1
    substring = num[i:i+13]
    for k in substring:
        product*=int(k)
    if product>greatest_product:
        greatest_product=product
print(greatest_product)
```
