## Initial approach

* start from smallest number
* complete all iterations for the smallest
* then move to next number upto 100
* max iteration of any number is 100//number
* record numbers into a list to avoid duplication
* count list length

```python
%%time

def count_ways(total, smallest):
    if total == 0:
        return 1
    if total < smallest:
        return 0
    ways = 0
    for i in range(smallest, total + 1):
        ways += count_ways(total - i, i)
    return ways

print("Result:", count_ways(100, 1) - 1)
```

**Note: This implementation is significantly slow and should be optimized in future iterations.**
