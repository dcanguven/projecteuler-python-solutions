## Initial approach

* You start with 1 pound and always bet the same fraction of your money.
* If the coin shows heads, your money increases by that fraction, and if it shows tails, it decreases by that fraction.
* After 1000 flips, your final money depends only on how many heads you got.
* We choose the fraction so that the required number of heads to reach one billion is as small as possible.
* The optimal fraction makes the target reachable with at least 602 heads.
* So the final answer is the probability of getting 602 or more heads out of 1000 fair coin flips.

```python
import math
from decimal import Decimal, getcontext

def solve():
    getcontext().prec = 50

    n = 1000
    required_heads = 602

    prob = Decimal(0)
    two_power = Decimal(2) ** n

    for k in range(required_heads, n + 1):
        prob += Decimal(math.comb(n, k)) / two_power

    return round(prob, 12)
```

```python
%%time
solution = solve()
print(f"{solution:.12f}")
```
