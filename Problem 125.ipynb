## Initial approach

* a number is valid if it is a palindrome and equals a sum of consecutive squares
* limit: 10^8
* try all possible square sequences
* if the total is palindrome, keep it
* add them together for the answer

```python
def is_palindrome(x):
    s = str(x)
    return s == s[::-1]

def palindromic_square_sums(limit=10**8):
    squares = [i*i for i in range(1, int(limit**0.5)+1)]
    results = set()
    n = len(squares)
    for i in range(n):
        total = squares[i] + squares[i+1] if i+1 < n else 0
        j = i + 2
        while total < limit and j <= n:
            if total >= 0 and is_palindrome(total):
                results.add(total)
            if j == n:
                break
            total += squares[j]
            j += 1
    return sum(results)

print(palindromic_square_sums())
```
