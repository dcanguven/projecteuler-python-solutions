## Initial approach

* count 20-digit numbers, no leading zero
* ensure every 3 consecutive digits sum ≤ 9
* use dp on last two digits only
* dp[a][b] = ways ending with a,b
* start with first digit 1–9, second 0–9
* next digit c valid if a+b+c ≤ 9
* update new[b][c] += dp[a][b]
* after 20 digits, sum all states for result

```python
dp = [[0]*10 for _ in range(10)]
for d1 in range(1, 10):
    for d2 in range(0, 10):
        dp[d1][d2] += 1

for _ in range(3, 21):
    ndp = [[0]*10 for _ in range(10)]
    for a in range(10):
        row = dp[a]
        for b in range(10):
            cnt = row[b]
            if cnt == 0:
                continue
            limit = 9 - (a + b)
            if limit < 0:
                continue
            for c in range(limit + 1):
                ndp[b][c] += cnt
    dp = ndp

result = sum(sum(row) for row in dp)
print("Result:", result)
```
