## Initial approach

* we want to count all n where n and n plus 1 have the same number of positive divisors
* the range is from 2 up to ten million, so we need divisor counts up to 10^7
* we build an array where each position holds how many divisors that number has
* for each i from 1 to 10^7 we walk through its multiples and increase their divisor counts
* this works because every divisor i contributes to all numbers that are multiples of i
* after we fill the array we loop from 2 to 10^7 minus 1 and compare d(n) with d(n+1)
* every time the divisor counts are equal we increase our answer
* at the end we print the total number of such n

```python
%%time

limit = 10**7
div_count = [0] * (limit + 1)

for i in range(1, limit + 1):
    for j in range(i, limit + 1, i):
        div_count[j] += 1

result = 0
for n in range(2, limit):
    if div_count[n] == div_count[n + 1]:
        result += 1

print("Result:", result)
```
