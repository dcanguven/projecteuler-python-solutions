## Initial approach

* we want numbers below 250 that are not divisible by any prime square like 4, 9, 25, 49 etc
* a number is squarefree if for every prime p, p² does not divide that number
* to test this, for a given n we can check all integers p from 2 up to sqrt(n)
* if for any p we find that p² divides n exactly, n is not squarefree
* if we finish checking all p up to sqrt(n) with no p² dividing n, then n is squarefree
* we repeat this test for every n from 1 up to 249 since the question says “below 250”
* we count how many of these n pass the squarefree test and that count is our final answer

```python
import math

def is_squarefree(n):
    p = 2
    while p * p <= n:
        if n % (p * p) == 0:
            return False
        p += 1
    return True

limit = 250
count = 0

for n in range(1, limit):
    if is_squarefree(n):
        count += 1
```

```python
%%time
result = count
print("Result:", result)
```
