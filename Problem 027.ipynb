## Initial approach

First define a function to see if number is prime or not

```python
def is_prime(num):
    if num<2: 
        return False
    if num==2:
        return True
    if num%2==0: 
        return False
    for i in range(3, int(num**0.5)+1):
        if num%i == 0:
            return False
    return True
```

Store all primes into primes list

```python
primes = list()
for i in range(2,1001):
    if is_prime(i):
        primes.append(i)
```

* Iterate through all numbers starting with n=0 with n + an + b formulation.
* if result is prime, increase n
* find the longest combination

```python
max_length = 0
product = 0

for a in range(-999, 1000):
    for b_prime in primes:
        for b in (b_prime, -b_prime):
            n = 0
            while True:
                value = n * n + a * n + b
                if is_prime(value):
                    n += 1
                else:
                    break
            if n > max_length:
                max_length = n
                best_a = a
                best_b = b
```

```python
print("Best a:", best_a)
print("Best b:", best_b)
print("a * b:", best_a * best_b)
print("Length of longest consecutive primes:", max_length)
```
