## Initial approach

Lets try to find first example manually

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
```

Check if the number has only have prime factors

```python
main_list = []
prev = None

for num in range(1, 1000):
    list1 = []
    for i in range(2, num + 1):
        if num % i == 0 and is_prime(i):
            if i not in list1:
                list1.append(i)
    if len(list1) == 2:
        if prev == num - 1:
            print(prev, num)
            break
        prev = num
    else:
        prev = None
```

Now try one with 3 consecutive numbers

```python
main_list = []

for num in range(1, 1000):
    list1 = []
    for i in range(2, num + 1):
        if num % i == 0 and is_prime(i):
            if i not in list1:
                list1.append(i)
    if len(list1) == 3:
        main_list.append(num)
        if len(main_list) >= 3:
            if main_list[-1] - main_list[-2] == 1 and main_list[-2] - main_list[-3] == 1:
                print(main_list[-3], main_list[-2], main_list[-1])
                break
    else:
        main_list = []
```

Now solve it for 4 consecutive numbers

```python
%%time

from tqdm import tqdm

main_list = []

for num in tqdm(range(1, 200000)):
    list1 = []
    for i in range(2, num + 1):
        if num % i == 0 and is_prime(i):
            if i not in list1:
                list1.append(i)
    if len(list1) == 4:
        main_list.append(num)
        if len(main_list) >= 4:
            if (main_list[-1] - main_list[-2] == 1 and
                main_list[-2] - main_list[-3] == 1 and
                main_list[-3] - main_list[-4] == 1):
                print(main_list[-4], main_list[-3], main_list[-2], main_list[-1])
                break
    else:
        main_list = []
```

This code works but is not optimized; it can be improved later for better performance.
