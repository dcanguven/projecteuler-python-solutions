* It is very slow because it checks too many numbers.
* It tries all possible numerators for each denominator.
* Code needs to be optimized.

```python
%%time

import math

best_numerator = 0
best_denominator = 1

for denominator in range(2, 1000001):
    numerator = (3 * denominator) // 7
    if (numerator * 7 == 3 * denominator):
        numerator -= 1
    if math.gcd(numerator, denominator) == 1:
        if numerator * best_denominator > best_numerator * denominator:
            best_numerator = numerator
            best_denominator = denominator

print("Result:", best_numerator)
```

* It directly finds the best numerator close to 3/7 using math.
* It is much faster because it skips unnecessary checks.
