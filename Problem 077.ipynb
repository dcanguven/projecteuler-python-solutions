## Initial approach

* this problem is very similar to previous problem
* start with the smallest number
* explore all the possible combinations that include this number
* move on to next number up to target
* each number can appear at most, target // number times
* ensure combinations are built in non-decreasing oder to aviod duplicates
* count the total number of valid combinations


```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def primes_up_to(n):
    return [i for i in range(2, n+1) if is_prime(i)]
```

```python
def count_ways_with_primes(total, smallest, primes):
    if total == 0:
        return 1
    if total < smallest:
        return 0
    ways = 0
    for p in primes:
        if p < smallest:
            continue
        if p > total:
            break
        ways += count_ways_with_primes(total - p, p, primes)
    return ways
```

```python
%%time
n = 2
while True:
    primes = primes_up_to(n)
    count = count_ways_with_primes(n, 2, primes)
    if count > 5000:
        print("Result:", n)
        break
    n += 1
```
