## Initial approach

Understand the example

```python
import math
print("Starting point:",169)
print(math.factorial(1)+math.factorial(6)+math.factorial(9))
print(2*(math.factorial(3)+math.factorial(6)))
print(math.factorial(1)+math.factorial(4)+math.factorial(5)+math.factorial(4))
print("Loop ends")
```

Precompute factorials of digits 0 to 9

```python
import math
factorials = dict()
for i in range(0,10):
    factorials[i]=math.factorial(i)

print(factorials)
```

Compute sum of factorials of digits

```python
def digit_factorial_sum(n):
    return sum(factorials[int(d)] for d in str(n))
```

Try the example again

```python
digit_factorial_sum(digit_factorial_sum(digit_factorial_sum(169)))
```

```python
def build_chain(n):
    seen = set()
    while n not in seen:
        seen.add(n)
        n = digit_factorial_sum(n)
    return len(seen)
```

Now, iterate

```python
%%time

count = 0
for i in range(1, 10**6):
    if build_chain(i) == 60:
        count += 1

print("Answer:", count)
```
