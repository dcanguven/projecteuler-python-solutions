## Initial approach

* pick n distinct letters and count permutations with exactly one ascent between neighbors
* the count over permutations depends only on n not on which letters we picked
* permutations of size n with exactly one ascent are given by the eulerian number A(n,1)
* use the closed form A(n,1) = 2**n - n - 1
* for each n the total strings are C(26,n) * A(n,1)
* scan n from 2 to 26 and keep the largest value
* the maximizer turns out to be n = 18 and we report the maximum value asked
* use only math.comb for binomials

```python
%%time
import math

def A_n1(n):
    return 2**n - n - 1

def f(n):
    return math.comb(26, n) * A_n1(n)

best_n = None
best_val = -1

for n in range(2, 27):
    val = f(n)
    if val > best_val:
        best_val = val
        best_n = n

print("Result:", best_val)
print("Best n:", best_n)
```
