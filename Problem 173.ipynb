## Initial approach

* a lamina is an outer L by L square with an inner hole of side L minus 2t where t is the border thickness
* the number of tiles used is L^2 minus (L minus 2t)^2 which equals 4t times (L minus t)
* for fixed t the smallest L is 2t plus 1 and it uses 4t times (t plus 1) tiles
* with a budget N we need 4t times (t plus 1) at most N and for each t the largest L is floor(N divided by 4t) plus t
* the count for a given t is max of zero and floor(N divided by 4t) plus t minus 2t
* sum this over t starting from 1 while 4t times (t plus 1) at most N
* set N to one million and compute the total

```python
def count_laminae(N):
    total = 0
    t = 1
    while 4 * t * (t + 1) <= N:
        maxL = N // (4 * t) + t
        total += max(0, maxL - 2 * t)
        t += 1
    return total
```

```python
%%time
result = count_laminae(10**6)
print("Result:", result)
```
