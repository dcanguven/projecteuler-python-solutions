## Initial approach

* use each pair only once because swapping the two numbers gives the same value
* ignore pairs where the product is a perfect square
* the useful cases are the ones where the smaller companion value stays between zero and one
* when that happens, the main power gets closer and closer to the next integer
* the number of leading nines depends on how fast that companion value shrinks
* compute the minimal exponent with logarithms
* sum all minimal exponents for valid pairs

```python
import math

def solve(limit, digits):
    total = 0
    target = digits * math.log(10)

    for p in range(1, limit + 1):
        for q in range(p + 1, limit - p + 1):
            root = math.isqrt(p * q)

            if root * root == p * q:
                continue

            companion = (math.sqrt(q) - math.sqrt(p)) ** 2

            if 0 < companion < 1:
                c = math.ceil(target / -math.log(companion))
                total += c

    return total
```

```python
%%time
result = solve(2011, 2011)
print("Result:", result)
```
