## Initial approach

* model the reachable region by looking at the highest possible path for each horizontal distance
* because fragments fly in every direction with the same speed, the outer boundary is a paraboloid
* the ground cuts this paraboloid at its maximum horizontal range
* rotate the boundary curve around the vertical axis to get the full 3D volume
* integrate circular slices from the ground up to the top height
* use standard floating point because the final answer only needs four decimals
* print the rounded volume directly

```python
import math

def solve():
    h = 100
    v = 20
    g = 9.81

    y_top = h + v * v / (2 * g)
    r_squared_at_ground = (v * v / g) ** 2 + 2 * h * v * v / g

    volume = math.pi * r_squared_at_ground * y_top / 2

    return volume
```

```python
%%time
result = solve()
print("Result:", f"{result:.4f}")
```
