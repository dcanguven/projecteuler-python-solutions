## Initial approach

* make all subsets using combinations
* pick two subsets with same size and no common elements
* compare their elements one by one
* if not all smaller and not all bigger, then they need testing
* count.

```python
from itertools import combinations

n = 12
total = 0

for r in range(2, n // 2 + 1):
    subsets = list(combinations(range(n), r))
    for i in range(len(subsets)):
        for j in range(i + 1, len(subsets)):
            A = subsets[i]
            B = subsets[j]
            if set(A).isdisjoint(B):
                A_sorted = sorted(A)
                B_sorted = sorted(B)
                all_less = True
                all_greater = True
                for x, y in zip(A_sorted, B_sorted):
                    if x >= y:
                        all_less = False
                    if x <= y:
                        all_greater = False
                if not all_less and not all_greater:
                    total += 1

print(total)
```
