## Initial approach

* write the series as x g1 + x^2 g2 +  and set it equal to an integer n, which forces a quadratic whose discriminant must be a perfect square
* this condition reduces to a pell type equation so valid x come in an infinite but structured sequence
* the pell solutions can be generated by a simple linear recurrence on auxiliary terms f_k
* the golden nuggets n are recovered from those f_k by n = f_k / 5  1
* seed the recurrence with four small solutions then iterate to get the first 30 nuggets
* sum these 30 integers to answer the problem
* use only the math library for 5

```python
%%time
import math

def first_30_nuggets_sum():
    f = [7, 14, 50, 97]
    for _ in range(30 - 4):
        f.append(7 * f[-2] - f[-4])
    sqrt5 = math.sqrt(5)
    nuggets = [int(x / sqrt5) - 1 for x in f]
    return sum(nuggets)

result = first_30_nuggets_sum()
print("Result:", result)
```
