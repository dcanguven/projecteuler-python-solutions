## Initial approach

* turn the series into x divided by 1 minus x minus x squared and set it equal to n
* solving for x gives a quadratic whose discriminant must be a perfect square for rational x
* this condition turns into a pell type relation that generates a clean integer sequence
* the nuggets line up as products of consecutive even indexed fibonacci numbers
* specifically the k th nugget equals F(2k) times F(2k+1)
* compute fibonacci numbers up to index 31 and take the product for k equal to 15
* the final number is large but fits easily in python integers

```python
def fib(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

k = 15
result = fib(2*k) * fib(2*k + 1)
print("Result:", result)
```
