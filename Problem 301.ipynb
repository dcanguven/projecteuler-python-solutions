## Initial approach

* the losing positions here follow a very neat binary pattern
* when comparing a number with its double and triple, the xor becomes zero exactly when adding the number to itself creates no carry
* that means the binary form of the number cannot contain two adjacent ones
* so the whole game question turns into a counting question about binary strings
* now count how many positive integers up to the given limit have binary digits with no consecutive ones
* this count follows the fibonacci pattern, because each valid binary prefix can end in zero freely or end in one only after a zero
* include zero first because the counting pattern is cleaner that way, then remove it at the end since the problem asks for positive integers

```python
def solve():
    a, b = 1, 2

    for _ in range(2, 31):
        a, b = b, a + b

    result = b
    print("Result:", result)
```
