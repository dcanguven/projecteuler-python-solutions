* So based on my research, I found this method:
* start from the smallest solution, b=3, total=4
* in each step use a formula to jump to next valid solution
* stop when total discs (n) is more than limit, 10^12
* find b

````python
b, n = 3, 4
limit = 10**12

while n <= limit:
    b, n = 3*b + 2*n - 2, 4*b + 3*n - 3

print("Result:", b)
````
