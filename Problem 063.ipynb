## Initial approach

Iterate few to see the case.  

First define a function to return digit number of a given number.

```python
def digits(num):
    return len(str(num))
```

Now exaggerate the iteration

```python
count = 0

for i in range(1,200):
    for k in range(1,200):
        if digits(i**k) == k:
            count+=1

print("Result: ", count)
```
