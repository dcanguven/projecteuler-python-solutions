## Initial approach

- dp[i] = number of ways to fill a row of length i.  
- pref[i] = sum of all dp values up to i (helps speed up).  
- Start: dp[0] = 1 (empty row has 1 way).  
- For each position i from 1 to n:  
  - Case 1: put a black square first → add dp[i-1].  
  - Case 2: place a red block of length ≥ 3.  
    - If it exactly fits the row → add 1.  
    - If space remains, force one black square, then use pref[i-4] for the rest.  
- Save total in dp[i].  
- Update prefix sum: pref[i] = pref[i-1] + dp[i].  
- At the end, dp[n] is the total number of ways.

```python
def count_ways(n):
    dp = [0]*(n+1)
    pref = [0]*(n+1)
    dp[0] = 1
    pref[0] = 1
    for i in range(1, n+1):
        val = dp[i-1]
        if i >= 3:
            val += 1
        if i >= 4:
            val += pref[i-4]
        dp[i] = val
        pref[i] = pref[i-1] + dp[i]
    return dp[n]

print("Result:", count_ways(7))
```
