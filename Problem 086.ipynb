## Initial approach

* lets start with sides up to max M value.
* for each M, try all other sides where side 1 + side 2 ≤ 2M
* calculate shortest surface walking path distance $\sqrt{M^2 + (a + b)^2}$
* if the parth is integer, count it
* keep adding and stop when total count goes over 1 million
* return smallest M where it happens

Here NumPy is faster because it reduces Python loops and uses C-speed math operations for the core of this problem:

* squaring,
* adding
* square roots.

```python
import numpy as np

def is_perfect_square(n):
    root = np.sqrt(n)
    return root == int(root)

def count_for_m(m):
    count = 0
    for ab in range(2, 2 * m + 1):
        path_squared = m * m + ab * ab
        if is_perfect_square(path_squared):
            count += ab // 2
    return count
```

```python
%%time

M = 0
total = 0

while total <= 1_000_000:
    M += 1
    total += count_for_m(M)

print("Result:", M)
```
