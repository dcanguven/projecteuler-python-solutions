## Initial approach

* this problem is similar with problem 112.
* counting upto 10^100 is not possible.
* we will use combination rather than counting

```python
from math import comb

increasing = comb(100 + 9, 9) - 1
decreasing = comb(100 + 10, 10) - (100 + 1)
```

* comb(100 + 9, 9) -1 : this counts all increasing numbers upto 100 digits, substract 1 to remove all zero case
* comb(100 + 10, 10) - (100 + 1): counts all decreasing, substract to remove over-counted cases (leading and all zeros)

```python
print("Result:", increasing + decreasing - (100*9))
```
