## Initial approach

* write a pythagorean triple with euclid’s formula and 
* note the outer square side is a+b and the hole side is |b−a|
* the tiling works exactly when the big side is an integer multiple of the hole side
* for primitive triples this reduces to the simple test c divisible by |b−a|
* if a primitive triple passes the test then all of its integer multiples also work
* generate primitive triples by coprime m>n with opposite parity and perimeter 2m(m+n)
* for each valid primitive add how many multiples fit under the perimeter limit

```python
%%time

import math

limit = 100_000_000
res = 0
m = 2
while 2*m*(m+1) < limit:
    for n in range(1, m):
        if ((m - n) & 1) == 1 and math.gcd(m, n) == 1:
            a = m*m - n*n
            b = 2*m*n
            c = m*m + n*n
            diff = abs(b - a)
            if c % diff == 0:
                p0 = a + b + c
                res += (limit - 1) // p0
    m += 1

print("Result:", res)
```
