## Initial approach

* look for k-digit n where rotating right once gives R = mn for some m in 1..9
* write n = 10q + d with last digit d in 1..9 and q = n/10
* equation becomes (10m  1)q = d(10^{k1}  m) so q must be an integer
* for each k = 2..100, m = 1..9, d = 1..9 compute q and n, then verify R = mn
* keep only k-digit n and sum them once each
* take the sum mod 100000 to get the last 5 digits

```python
def solve():
    L = 100000
    total = 0
    seen = set()
    for k in range(2, 101):
        p = 10**(k-1)
        for m in range(1, 10):
            s = 10*m - 1
            t = p - m
            for d in range(1, 10):
                num = d * t
                if num % s != 0:
                    continue
                q = num // s
                n = 10*q + d
                if not (p <= n < 10*p):
                    continue
                R = d*p + q
                if R == m * n:
                    if n not in seen:
                        seen.add(n)
                        total = (total + n) % L
    return total

result = solve()
print("Result:", result)
```
