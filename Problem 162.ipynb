## Initial approach

* we need hex numbers up to 16 digits long with no leading zero and each of 0, 1, A appearing at least once
* count by length n and sum for n from 1 to 16 since at most sixteen digits
* use inclusion exclusion on the forbidden set among {0,1,A} while respecting the leading digit rule
* for a forbidden set S, positions after the first allow 16|S| choices, the first allows (16|S|) minus 1 if zero is allowed
* this gives for each n the formula 1516^(n1)  (15^n + 2815^(n1)) + (214^n + 1314^(n1))  13^n
* sum these counts over n=1..16 to get the total in decimal
* finally convert the total to uppercase hexadecimal with no prefix and print it

```python
def count_for_n(n):
    return 15*(16**(n-1)) - (15**n + 28*(15**(n-1))) + (2*(14**n) + 13*(14**(n-1))) - 13**n
```

```python
%%time

total = sum(count_for_n(n) for n in range(1, 17))
result = format(total, 'X')
print("Result:", result)
```
