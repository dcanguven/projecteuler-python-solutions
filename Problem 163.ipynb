## Initial approach

* size 1 triangle has 16 triangles total
* size 2 triangle has 104 triangles total
* each new size adds smaller triangles inside that overlap in a recursive pattern
* we can express T(n) (total triangles) as
T(n) = T(n  1) + (n  1)  (n  1)  16 + 6  (n  1)  (n  2)
* use a loop from 1 to 36 to compute recursively
* print T(36)

```python
%%time

def T(n):
    if n == 1:
        return 16
    if n == 2:
        return 104
    t = [0]*(n+1)
    t[1], t[2] = 16, 104
    for i in range(3, n+1):
        t[i] = t[i-1] + (i-1)*(i-1)*16 + 6*(i-1)*(i-2)
    return t[n]

result = T(36)
print("Result:", result)
```
