## Initial approach

Try the function from Problem 02

```python
def fibonacci(limit):
    k0 = 0
    k1 = 1
    num = 1
    list1 = [0]
    while num<limit:
        list1.append(num)
        num = k0 + k1
        k0 = k1
        k1 = num
    return list1

# example:
# fib number upto 100:
fibonacci(100)
```

We need a function that returns number of digits of a given number

```python
def digit_count(n):
    return len(str(abs(n)))
```

Try last element of fibonacci(100) function

```python
fibonacci(100)[-1]
```

```python
digit_count(fibonacci(100)[-1])
```

Lets combine both approach:

```python
def check_fib_digits(digits):
    a, b = 1, 1
    index = 2
    while digit_count(b) < digits:
        a, b = b, a + b
        index += 1
    return index
```

```python
check_fib_digits(1000)
```

## Clean and tested code

```python
%%time

def check_fib_digits(digits):
    a, b = 1, 1
    index = 2
    while len(str(b)) < digits:
        a, b = b, a + b
        index += 1
    return index

print(check_fib_digits(1000))
```
