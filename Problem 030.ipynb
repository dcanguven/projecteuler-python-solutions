## Initial approach

Generate a function which return True, if total is equal to given number.

```python
def is_total_equal_4(num):
    num_as_list = list(str(num))
    total = 0
    for i in num_as_list:
        total += int(i)**4
    if total == num:
        return True
    else:
        return False
```

```python
print("Check True: ", is_total_equal_4(1634))
print("Check False: ", is_total_equal_4(1633))
```

Good, lets modify the function for 5th power.

```python
def is_total_equal_5(num):
    num_as_list = list(str(num))
    total = 0
    for i in num_as_list:
        total += int(i)**5
    if total == num:
        return True
    else:
        return False
```

Now, iterate:

```python
%%time

result = 0
for i in range(2,1000000):
    if is_total_equal_5(i):
        result += i

result
```

## Clean and tested code

```python
%%time

def is_total_equal_5_V2():
    return sum(i for i in range(10, 354295) if i == sum(int(d)**5 for d in str(i)))

print(is_total_equal_5_V2())
```
