## Initial approach

* every number is already a two-digit repunit in base one less than itself
* so we only need to generate repunits with at least three digits
* for each base, repeatedly build the value by multiplying by the base and adding one
* store all values in a set to avoid duplicates
* include 1 manually because it is counted in the problem
* stop a base once the smallest three-digit repunit reaches the limit
* sum all unique strong repunits below the limit

```python
def solve(limit):
    values = {1}

    base = 2

    while base * base + base + 1 < limit:
        value = 1 + base + base * base

        while value < limit:
            values.add(value)
            value = value * base + 1

        base += 1

    return sum(values)
```

```python
%%time
assert solve(50) == 171
result = solve(10 ** 12)
print("Result:", result)
```
