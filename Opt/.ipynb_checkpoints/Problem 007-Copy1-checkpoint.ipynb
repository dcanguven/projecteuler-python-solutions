## Thought Process

1. Use the `is_prime` function to check whether a number is prime.
2. Create an empty dictionary instead of a list.
3. In each loop iteration, check the length of the dictionary.
4. Stop the loop once the dictionary contains **10001** prime numbers.
5. Print the **last inserted value**, which will be the 10001st prime number.
6. Store each prime number in the dictionary with the key as its index (e.g., `1: 2`, `2: 3`, ..., `10001: ???`).

## Initial approach

```python
def is_prime(n):
    for i in range(2,n//2+1):
        if n%i==0:
            return False
    return True
```

```python
%%time

primes=dict()
i = 2
while len(primes)<10001:
    if is_prime(i):
        primes[len(primes)+1]=i
    i+=1

primes[10001]
```

## Clean and tested code

```python
%%time

import math

def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(math.isqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

primes = dict()
index = 1
i = 2
while index <= 10001:
    if is_prime(i):
        primes[index] = i
        index += 1
    i += 1

print(primes[10001])
```
