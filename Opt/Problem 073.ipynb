## Initial approach

* Loop through all denominators from 2 to 12000.
* Find the smallest numerator so that the fraction is just above 1/3.
* Find the largest numerator so that the fraction is just below 1/2.
* Check if the numerator and denominator are relatively prime
* if yes, count.

Note: Relatively prime mentioned in problem 69, problem 70 and 72.

```python
%%time
import math

count = 0

for denominator in range(2, 12001):
    min_numerator = denominator // 3 + 1
    max_numerator = (denominator - 1) // 2
    for numerator in range(min_numerator, max_numerator + 1):
        if math.gcd(numerator, denominator) == 1:
            count += 1

print("Result:", count)
```
