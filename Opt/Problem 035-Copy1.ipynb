## Initial approach

* int-->str, change order of numbers, str-->int, store them
* check if all prime or not
* use a counter for "True"s.

```python
def is_prime(n):
    if n < 2: return False
    if n == 2: return True
    if n % 2 == 0: return False
    for i in range(3, int(n**0.5)+1, 2):
        if n % i == 0: return False
    return True
```

```python
num = 197
if num<1000:
    num=str(num)
    num1 = num[0]+num[1]+num[2]
    num2 = num[2]+num[0]+num[1]
    num3 = num[1]+num[2]+num[0]
    num1=int(num1)
    num2=int(num2)
    num3=int(num3)
    if is_prime(num1)==is_prime(num2)==is_prime(num3):
        print("true, continue!")
```

Now we can built a function for all options below 1M.

```python
def is_circular_prime(num):
    if num<1000:
        num=str(num)
        num1 = num[0]+num[1]+num[2]
        num2 = num[2]+num[0]+num[1]
        num3 = num[1]+num[2]+num[0]
        num1=int(num1)
        num2=int(num2)
        num3=int(num3)
        if is_prime(num1)==True and is_prime(num2)==True and is_prime(num3)==True:
            return True
    elif num<10000:
        num=str(num)
        num1 = num[0]+num[1]+num[2]+num[3]
        num2 = num[1]+num[2]+num[3]+num[0]
        num3 = num[2]+num[3]+num[0]+num[1]
        num4 = num[3]+num[0]+num[1]+num[2]
        num1=int(num1)
        num2=int(num2)
        num3=int(num3)
        num4=int(num4)
        if is_prime(num1)==True and is_prime(num2)==True and is_prime(num3)==True and is_prime(num4)==True:
            return True
    elif num<100000:
        num=str(num)
        num1 = num[0]+num[1]+num[2]+num[3]+num[4]
        num2 = num[1]+num[2]+num[3]+num[4]+num[0]
        num3 = num[2]+num[3]+num[4]+num[0]+num[1]
        num4 = num[3]+num[4]+num[0]+num[1]+num[2]
        num5 = num[4]+num[0]+num[1]+num[2]+num[3]
        num1=int(num1)
        num2=int(num2)
        num3=int(num3)
        num4=int(num4)
        num5=int(num5)
        if is_prime(num1)==True and is_prime(num2)==True and is_prime(num3)==True and is_prime(num4)==True and is_prime(num5)==True:
            return True
    elif num<1000000:
        num=str(num)
        num1 = num[0]+num[1]+num[2]+num[3]+num[4]+num[5]
        num2 = num[1]+num[2]+num[3]+num[4]+num[5]+num[0]
        num3 = num[2]+num[3]+num[4]+num[5]+num[0]+num[1]
        num4 = num[3]+num[4]+num[5]+num[0]+num[1]+num[2]
        num5 = num[4]+num[5]+num[0]+num[1]+num[2]+num[3]
        num6 = num[5]+num[0]+num[1]+num[2]+num[3]+num[4]
        num1=int(num1)
        num2=int(num2)
        num3=int(num3)
        num4=int(num4)
        num5=int(num5)
        num6=int(num6)
        if is_prime(num1)==True and is_prime(num2)==True and is_prime(num3)==True and is_prime(num4)==True and is_prime(num5)==True and is_prime(num6)==True:
            return True
    else:
        return False
```

Now, iteration from 100. Counter will start from 13 since there are 13 primes below 100.

```python
%%time
counter = 13
for i in range(100,1_000_000):
    if is_circular_prime(i):
        counter+=1

print(counter)
```

## Optimize the function

```python
def is_circular_prime(num):
    num = str(num)
    for i in range(len(num)):
        rotated = int(num[i:] + num[:i])
        if not is_prime(rotated):
            return False
    return True
```

```python
%%time
counter = 13
for i in range(100,1_000_000):
    if is_circular_prime(i):
        counter+=1

print(counter)
```

## Optimize the iteration (use only odd numbers)

```python
%%time
counter = 13
for i in range(101,1_000_000,2):
    if is_circular_prime(i):
        counter+=1

print(counter)
```
