## Initial approach

* iterate a and b from 1000 down to 100.
* check if a x b is a palindrome or not by using is_palindromeV1 function.
* if it's a palindrome and greater than the previous one, store it as a new maximum.

```python
%%time

def is_palindromeV1(n):
    n = str(n)
    for i in range(len(n) // 2):
        if n[i] != n[-1 - i]:
            return False
    return True

product = 1
first_factor = 1
second_factor = 1

for a in range(1000, 99, -1):
    for b in range(1000, 99, -1):
        if is_palindromeV1(a * b):
            if a * b > product:
                product = a * b
                first_factor = a
                second_factor = b

print(f"{first_factor} x {second_factor} = {product}")
```

## Clean and tested code

```python
%%time

def is_palindrome(n):
    s = str(n)
    return s == s[::-1]

max_product = 0
first_factor = 0
second_factor = 0

for a in range(999, 99, -1):
    for b in range(a, 99, -1):  # avoid redundant pairs
        product = a * b
        if product <= max_product:
            break  # no larger product possible
        if is_palindrome(product):
            max_product = product
            first_factor = a
            second_factor = b

print(f"{first_factor} x {second_factor} = {max_product}")
```
