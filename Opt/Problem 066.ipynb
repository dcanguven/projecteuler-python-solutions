## Initial approach

* Loop from D = 2 to 1000
* skip perfect squares
* solve x^2 - D y^2 = 1 for minimal solution
* record maximal x and its D
* return that D

```python
def minimal_solution_x(D):
    m = 0
    d = 1
    a0 = int(math.sqrt(D))
    if a0 * a0 == D:
        return 0
    a = a0
    num1, num2 = 1, a
    den1, den2 = 0, 1
    while num2 * num2 - D * den2 * den2 != 1:
        m = d * a - m
        d = (D - m * m) // d
        a = (a0 + m) // d
        num1, num2 = num2, a * num2 + num1
        den1, den2 = den2, a * den2 + den1
    return num2
```

```python
import math

max_x = 0
result_D = 0

for D in range(2, 1001):
    x = minimal_solution_x(D)
    if x > max_x:
        max_x = x
        result_D = D

print("Result:",result_D)
```
