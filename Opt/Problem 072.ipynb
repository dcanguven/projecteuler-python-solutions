What is Totient Function

- It counts numbers that are relatively prime to n.
- Relatively prime means GCD is 1.

Why We Need Totient Here (Problem 72)

- We count reduced fractions.
- Denominator can be up to 1,000,000.
- For each denominator, we count numerators with GCD 1.
- Totient gives this count directly.
- It is much faster than checking all numbers.

Simple Summary

- Without totient, as we tried: check all numbers, very slow.
- With totient: count directly, very fast.

```python
%%time

limit = 1000000
totients = [i for i in range(limit + 1)]

for i in range(2, limit + 1):
    if totients[i] == i:
        for j in range(i, limit + 1, i):
            totients[j] -= totients[j] // i

print("Result",sum(totients[2:]))
```
