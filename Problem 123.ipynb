## Initial approach

* start from odd n = 1, 3, 5...
* generate the n-th prime number
* compute the remainder = 2 * n * pn
* stop and return n when remainder > 10^10

* start by defining prime

```python
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True
```

* now generate n-th prime number

```python
def generate_primes():
    num = 2
    while True:
        if is_prime(num):
            yield num
        num += 1
```

* rest:

```python
%%time

limit = 10**10
n = 1
prime_gen = generate_primes()

while True:
    prime = next(prime_gen)
    if n % 2 == 1:
        remainder = 2 * n * prime
        if remainder > limit:
            print(n)
            break
    n += 1
```
