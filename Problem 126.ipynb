## Initial approach

* start with a cuboid (box) of size x × y × z
* each new layer wraps around the previous one
* count how many cubes are needed for each layer
* find how many different cuboids produce a given number of cubes for some layer

### defaultdict library:
* it automatically creates a default value if the key does not exist
* you don’t need to check if the key exists before adding or updating
* it saves time and avoids errors like KeyError

```python
from collections import defaultdict

def cube_layer_count(x, y, z, layer):
    base = 2 * (x * y + y * z + z * x)
    extra = 4 * (layer - 1) * (x + y + z + layer - 2)
    return base + extra

target = 1000
limit = 1000000
cube_counts = defaultdict(int)

for x in range(1, 100):
    for y in range(x, 100):
        for z in range(y, 100):
            layer = 1
            cubes = cube_layer_count(x, y, z, layer)
            while cubes < limit:
                cube_counts[cubes] += 1
                layer += 1
                cubes = cube_layer_count(x, y, z, layer)

for n in range(limit):
    if cube_counts[n] == target:
        print(n)
        break
```
