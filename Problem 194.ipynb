## Initial approach

* we see the A and B pieces as small fixed graphs, glued in a long strip by matching their vertical edges
* the only communication between units is through the four vertices on each vertical edge, so each possible colouring of these four vertices is treated as a state
* for a symbolic number of colours c, we count how many valid colourings of a single A piece map each left state to each right state, this gives a transition matrix for A whose entries are polynomials in c
* we do the same for B, getting a transition matrix for B, again with polynomial entries in c
* a whole chain with a copies of A and b copies of B is then described by multiplying these transition matrices in the correct order and summing over all possible start and end states
* at the very end we substitute c = 1984 into the polynomial entries, reduce everything modulo 10^8, and perform the matrix multiplications in modular arithmetic to get the last 8 digits
