## Initial approach

Pythagorean theorem: a^2 + b^2 = c^2   
where a and b are the legs, and c is the hypotenuse

```python
%%time

a = 3
b = 4
c = 5
list_of_triangles = list()

def check_int_condition(a,b,c):
    if a**2+b**2==c**2:
        return True

for a in range(1,500):
    for b in range(1,500):
        for c in range(1,500):
            if check_int_condition(a,b,c):
                list_of_triangles.append([a,b,c])

len(list_of_triangles)
```

This code takes too long, even though we are only looking for first 500 numbers. 

* use Euclids formula to generate Pythagorean triples.
* loop over values
* calculate perimeter
* add all multiples of p up to 1.500.000
* count how many times each perimeter value appears
* count how many perimeter values appear exactly once

```python
import math

limit = 1_500_000
counts = [0] * (limit + 1)

for m in range(2, int(math.sqrt(limit // 2)) + 1):
    for n in range(1, m):
        if (m - n) % 2 == 1 and math.gcd(m, n) == 1:
            a = m * m - n * n
            b = 2 * m * n
            c = m * m + n * n
            p = a + b + c

            while p <= limit:
                counts[p] += 1
                p += a + b + c

result = 0
for i in range(len(counts)):
    if counts[i] == 1:
        result += 1

print("Result:", result)
```
