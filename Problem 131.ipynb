## Initial approach

* make a sieve function to mark all primes up to 1 million
* create a boolean list isprime to check primality fast
* formula for candidate primes: p = 3a² + 3a + 1
* loop over a = 1,2,3,… until p ≥ 1 million
* if p is prime, increase count
* when loop ends, print count of such primes

```python
def sieve(n):
    s = bytearray(b"\x01")*(n+1)
    s[0:2] = b"\x00\x00"
    p = 2
    while p*p <= n:
        if s[p]:
            step = p
            start = p*p
            s[start:n+1:step] = b"\x00"*(((n - start)//step) + 1)
        p += 1
    return s

limit = 1_000_000
isprime = sieve(limit)
count = 0
a = 1
while True:
    p = 3*a*a + 3*a + 1
    if p >= limit:
        break
    if isprime[p]:
        count += 1
    a += 1
print("Result:", count)
```
