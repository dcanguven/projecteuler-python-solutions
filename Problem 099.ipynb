* it's clear that it cannot be solved by a traditional method
* lets try a rough method to see if we will be lucky or not
* as a start, calculate and compare digit counts

* example from question:

```python
import math

base = 632382
exponent = 518061
digit_count = math.floor(exponent * math.log10(base)) + 1

print(digit_count)
```

* Now, try this approach on txt file.

```python
with open("0099_base_exp.txt") as file:
    lines = [line.strip() for line in file if line.strip()]

max_digit_count = 0
max_lines = []

for index, line in enumerate(lines):
    base, exponent = map(int, line.split(","))
    digit_count = math.floor(exponent * math.log10(base)) + 1

    if digit_count > max_digit_count:
        max_digit_count = digit_count
        max_lines = [(index + 1, digit_count, line)]
    elif digit_count == max_digit_count:
        max_lines.append((index + 1, digit_count, line))

for line_number, digit, content in max_lines:
    print(f"Line {line_number} has {digit} digits  {content}")
```

* We were not lucky enough

* we use logarithms to compare these large numbers

```python
base1, exp1 = 804712, 508889
base2, exp2 = 895447, 504922

val1 = exp1 * math.log(base1)
val2 = exp2 * math.log(base2)

if val1 > val2:
    print("804712^508889 is larger")
    print("Result: Line 172")
else:
    print("895447^504922 is larger")
    print("Result: Line 709")
```
