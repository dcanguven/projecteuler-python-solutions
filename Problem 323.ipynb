## Initial approach

* think bit by bit instead of tracking full integers
* one bit is still zero after several random numbers only if it was zero every time
* the process finishes when all 32 bits have appeared as 1 at least once
* use the tail-sum idea for expected waiting time
* for each step, compute the probability that at least one bit is still missing
* sum these probabilities until the remaining terms are tiny
* floating point is enough because the answer needs 10 decimals

```python
def solve(bits):
    expected = 0.0
    n = 0

    while True:
        probability_done = (1 - 0.5 ** n) ** bits
        probability_not_done = 1 - probability_done

        if probability_not_done < 1e-15:
            break

        expected += probability_not_done
        n += 1

    return expected
```

```python
%%time
result = solve(32)
print("Result:", f"{result:.10f}")
```
