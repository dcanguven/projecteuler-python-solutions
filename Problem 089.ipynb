## Initial approach

* count characters in original roman number
* calculate define numerical number
* then convert it to roman number
* count new characters
* calculate difference

* Lets try "16" from example.
* first, generate a dictionary from values

I = 1  
V = 5  
X = 10  
L = 50  
C = 100  
D = 500  
M = 1000  

```python
values = {"I":1, "V":5, "X":10, "L":50, "C":100, "D": 500, "M":1000}
```

```python
examples = ["IIIIIIIIIIIIIIII", "VIIIIIIIIIII", "VVIIIIII", "XIIIIII", "VVVI", "XVI"]

def roman_to_int(roman):
    total = 0
    for char in roman:
        total += values[char]
    return total

numerical_numbers = [roman_to_int(r) for r in examples]
print(numerical_numbers)
```

Now, define these numbers in Roman form.

```python
def int_to_roman(number):
    result = ""
    while number >= 1000:
        result += "M"
        number -= 1000
    if number >= 900:
        result += "CM"
        number -= 900
    if number >= 500:
        result += "D"
        number -= 500
    if number >= 400:
        result += "CD"
        number -= 400
    while number >= 100:
        result += "C"
        number -= 100
    if number >= 90:
        result += "XC"
        number -= 90
    if number >= 50:
        result += "L"
        number -= 50
    if number >= 40:
        result += "XL"
        number -= 40
    while number >= 10:
        result += "X"
        number -= 10
    if number == 9:
        result += "IX"
        number -= 9
    if number >= 5:
        result += "V"
        number -= 5
    if number == 4:
        result += "IV"
        number -= 4
    while number >= 1:
        result += "I"
        number -= 1
    return result
```

```python
int_to_roman(16)
```

* Now calculate character savings from the list given.

```python
with open("0089_roman.txt") as file:
    examples = [line.strip() for line in file]

total_saved = 0
for roman in examples:
    value = roman_to_int(roman)
    minimal = int_to_roman(value)
    total_saved += len(roman) - len(minimal)

print("Result:", total_saved)
```
