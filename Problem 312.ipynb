## Initial approach

* first derive a direct count formula for the cycle count
* the count grows too fast, so only modular values are useful
* use the known small values to check the formula
* reduce the nested inputs using the period pattern modulo powers of 13
* compute the inside value with a smaller modulus first
* use that result as the next input modulus
* repeat until the final value modulo 13 to the power 8 is found

```python
def C(n, mod):
    if n == 1 or n == 2:
        return 1 % mod

    if n == 3:
        return 8 % mod

    exponent = (pow(3, n - 2) - 3) // 2
    return 8 * pow(12, exponent, mod) % mod

def solve():
    x1 = 10000 % (6 * 13 ** 2)
    x2 = C(x1, 6 * 13 ** 4)
    x3 = C(x2, 6 * 13 ** 6)
    x4 = C(x3, 13 ** 8)

    return x4

assert C(1, 10 ** 9) == 1
assert C(2, 10 ** 9) == 1
assert C(3, 10 ** 9) == 8
assert C(5, 10 ** 18) == 71328803586048
assert C(10000, 10 ** 8) == 37652224
assert C(10000, 13 ** 8) == 617720485
```

```python
%%time
result = solve()
print("Result:", result)
```
