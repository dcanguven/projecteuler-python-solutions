## Initial approach

* lets understand the case for n = 4

````python
list_of_solutions = list()

for x in range(1,500):
    for y in range(1,500):
        equation = (x**-1 + y**-1)**-1
        if equation == 4:
            list_of_solutions.append(tuple(sorted((x, y))))

print("list of solutions:", list(set(list_of_solutions)))
print("number of solutions:", len(list(set(list_of_solutions))))
````

* same approach for larger numbers:

%%time

        for n in range(1, 10**7):

                list_of_solutions.append(tuple(sorted((x, y))))
                if len(list_of_solutions) > 1000:

import math

list_of_solutions = []
n = 1

while True:
    list_of_solutions.clear()
    n2 = n * n
    r = int(math.isqrt(n2))
    for d in range(1, r + 1):
        if n2 % d == 0:
            a = d
            b = n2 // d
            if a <= b:
                x = n + a
                y = n + b
                list_of_solutions.append((x, y))
                if len(list_of_solutions) > 1000:
                    print(n)
                    raise SystemExit
            if a != b:
                a2 = b
                b2 = n2 // a2
                if a2 <= b2:
                    x = n + a2
                    y = n + b2
                    list_of_solutions.append((x, y))
                    if len(list_of_solutions) > 1000:
                        print("n value:", n)
                        raise SystemExit
    n += 1
