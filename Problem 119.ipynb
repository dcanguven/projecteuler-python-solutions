## Initial approach

* calculate digit sum
* generate powers
* add condition, if a^b == a
* sort the list

```python
def digit_sum(n):
    return sum(int(c) for c in str(n))

numbers = []

for a in range(2, 200):         
    for b in range(2, 30):       
        n = a ** b
        if digit_sum(n) == a:
            numbers.append(n)

numbers = sorted(numbers)
```

* check conditions in question (a2 and a10)

```python
print("a2:", numbers[2-1])
print("a10:", numbers[10-1])
```

* print a30

```python
print("Result (a30):", numbers[30-1])
```
