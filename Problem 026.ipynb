Easiest approach is:

* define if there is any recurring cycle
* if yes, find length
* iterate 1000 times

```python
def has_recurring_cycle(d):
    remainders = set()
    numerator = 1
    while numerator != 0:
        if numerator in remainders:
            return True
        remainders.add(numerator)
        numerator = (numerator * 10) % d
    return False
```

```python
def recurring_cycle_length(d):
    remainders = {}
    numerator = 1
    position = 0
    while numerator != 0 and numerator not in remainders:
        remainders[numerator] = position
        numerator = (numerator * 10) % d
        position += 1
    return 0 if numerator == 0 else position - remainders[numerator]
```

### iteration:

```python
%%time

max_len = 0
result = 0

for i in range(2, 1000):
    if has_recurring_cycle(i):
        length = recurring_cycle_length(i)
        if length > max_len:
            max_len = length
            result = i

print(f"num: {result}")
print(f"max cycle length: {max_len}")
```
