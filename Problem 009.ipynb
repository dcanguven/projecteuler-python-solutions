## Clean and tested code

```python
%%time

total = 1000

for a in range(1, total // 3):
    for b in range(a + 1, (total - a) // 2):
        c = total - a - b
        if a * a + b * b == c * c:
            print(a * b * c)
            break
    else:
        continue
    break
```
