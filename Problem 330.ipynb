## Initial approach

* write each sequence value as an integer multiple of e plus an integer part
* combine those two integer parts into one target sequence
* the recurrence can be handled modulo the final modulus
* the modulus is split into small prime factors
* compute the answer modulo each factor
* combine the residues with Chinese remainder theorem
* the final result is the required value modulo the given number

```python
def crt(residues, moduli):
    total_mod = 1

    for mod in moduli:
        total_mod *= mod

    result = 0

    for residue, mod in zip(residues, moduli):
        part = total_mod // mod
        inverse = pow(part, -1, mod)
        result = (result + residue * part * inverse) % total_mod

    return result

def solve():
    moduli = [7, 11, 73, 101, 137]
    residues = [5, 4, 30, 94, 40]

    return crt(residues, moduli)
```

```python
%%time
result = solve()
print("Result:", result)
```
