## Initial approach

* let the triangle have integer sides a ≤ b ≤ c and perimeter ≤ 100 million
* use the angle bisector theorem to express BD in terms of a, b, c
* write the ratio area(ABC) / area(ABD) using base and height, the height cancels
* this reduces the condition to a simple divisibility relation between side lengths
* after simplification, valid triangles fall into a few parametric families
* generate triangles from these parameter formulas instead of brute forcing all triples
* for each generated triangle check the perimeter bound
* count all valid triangles

```python
from math import gcd, isqrt

LIMIT = 100_000_000

def solve():
    count = 0

    m = 2
    while True:
        if 2 * m * (m + 1) > LIMIT:
            break
        for n in range(1, m):
            if gcd(m, n) != 1:
                continue
            a = m * m - n * n
            b = 2 * m * n
            c = m * m + n * n
            p = a + b + c
            if p > LIMIT:
                continue
            k = 1
            while k * p <= LIMIT:
                count += 1
                k += 1
        m += 1

    print("Result:", count)
```

```python
%%time
solve()
```
