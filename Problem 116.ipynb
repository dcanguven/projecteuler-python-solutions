## Initial approach

* count how many ways we can replace black tilers (length 1) with colored tiles (length 2,3 or 4)
* for each color, place k tiles of that length on a row of length 50
* for each color, count ways
* number of ways = C(50-k*length+k,k)
* sum all valid k
* add resuls for all colors

```python
from math import comb

def ways_for_length(n, L):
    total = 0
    for k in range(1, n // L + 1):
        total += comb(n - k*L + k, k)
    return total

n = 50

red = ways_for_length(n, 2)
green = ways_for_length(n, 3)
blue = ways_for_length(n, 4)
result = red + green + blue

print("Result:", result)
```
