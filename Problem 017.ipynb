## Initial approach

* convert numbers to words
* count letters

```python
def number_to_words(num):
    if not 0 <= num <= 9999999:
        return "Number out of range"

    ones = {
        0: "zero", 1: "one", 2: "two", 3: "three",
        4: "four", 5: "five", 6: "six", 7: "seven",
        8: "eight", 9: "nine"
    }

    teens = {
        10: "ten", 11: "eleven", 12: "twelve", 13: "thirteen",
        14: "fourteen", 15: "fifteen", 16: "sixteen",
        17: "seventeen", 18: "eighteen", 19: "nineteen"
    }

    tens = {
        2: "twenty", 3: "thirty", 4: "forty", 5: "fifty",
        6: "sixty", 7: "seventy", 8: "eighty", 9: "ninety"
    }

    result = []

    if num >= 1000:
        result.append(ones[num // 1000] + "thousand")
        num %= 1000

    if num >= 100:
        result.append(ones[num // 100] + "hundred")
        num %= 100

    if num > 0 and len(result) > 0:
        result.append("and")

    if 10 <= num <= 19:
        result.append(teens[num])
    else:
        if num >= 20:
            result.append(tens[num // 10])
            num %= 10
        if 0 < num < 10:
            result.append(ones[num])
        elif len(result) == 0 and num == 0:
            result.append(ones[0])

    return "".join(result)
```

```python
%%time

full_string = ""
for i in range(1,1001):
    full_string += number_to_words(i)
len(full_string)
```

## Initial approach is clean, simple, and efficient enough for the given input size.
