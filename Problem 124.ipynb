## Initial approach

* define rad(n) and compute if for all n (from 1 to 100000)
* make mairs (n, rad(n))
* sort by rad(n)
* if equal, sort by n
* define E(k), k-th n in the sorted list
* find E(10000)

```python
def rad(n):
    result = 1
    x = n
    p = 2
    while p * p <= x:
        if x % p == 0:
            result *= p
            while x % p == 0:
                x //= p
        p += 1 if p == 2 else 2
    if x > 1:
        result *= x
    return result
```

```python
pairs = [(n, rad(n)) for n in range(1, 100001)]
pairs.sort(key=lambda x: (x[1], x[0]))
print("Result", pairs[9999][0])
```
