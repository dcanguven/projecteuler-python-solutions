## Initial approach

```python
%%time
def v1_solution(n):
    sum_square=0
    square_sum=0
    for i in range(1,n+1):
        sum_square+=i**2
    for i in range(1,n+1):
        square_sum+=i
    square_sum=square_sum**2
    difference = square_sum - sum_square
    return difference

v1_solution(100)
```

## Clean and tested code

```python
%%time

def v1_solution(n):
    sum_square = n * (n + 1) * (2 * n + 1) // 6
    square_sum = (n * (n + 1) // 2) ** 2
    return square_sum - sum_square

v1_solution(100)
```
