## Initial approach

* build all dart scores
* the last dart must be a double
* consider 1,2 or 3 dart checkouts with the last dart fixes as a double
* for the first one or two darts order doesn't matter
* sum of all darts must be less than 100
* count each distinct score set only ONCE

```python
from itertools import combinations_with_replacement

singles = [("S"+str(n), n) for n in range(1, 21)]
doubles = [("D"+str(n), 2*n) for n in range(1, 21)]
triples = [("T"+str(n), 3*n) for n in range(1, 21)]
bulls = [("SB", 25), ("DB", 50)]
```

```python
first_two = singles + doubles + triples + bulls + [("0", 0)]
last_darts = doubles + [("DB", 50)]
values_first = [v for _, v in first_two]
values_last = [v for _, v in last_darts]
```

```python
%%time

count = 0
for i, j in combinations_with_replacement(range(len(first_two)), 2):
    s = values_first[i] + values_first[j]
    for v in values_last:
        if s + v < 100:
            count += 1

print("Result:", count)
```
