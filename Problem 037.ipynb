## Initial approach

* start with an is_prime function
* int--> str to index --> int
* run is_prime function for each iteration
* iterate through 1M and see if the final count is 11. if not, try larger numbers
* start iteration from 10
* iterate through only odd numbers
* add number to final sum

```python
def is_prime(n):
    if n < 2: return False
    if n == 2: return True
    if n % 2 == 0: return False
    for i in range(3, int(n**0.5)+1, 2):
        if n % i == 0: return False
    return True
```

```python
def check_trunc_prime(num):
    num=str(num)
    for i in range(1, len(num)):
        if not is_prime(int(num[i:])) or not is_prime(int(num[:i])):
            return False
    return True
```

```python
%%time

count = 0
total = 0
for i in range(11, 1_000_000, 2):
    if check_trunc_prime(i):
        total += i
        count += 1
        if count == 11:
            print("iteration stops at: ",i)
            break
print("total numbers: ",count)
print("total sum: ",total)
```
