## Initial approach

```python
%%time

def sum_of_divisors(n):
    total = 1
    sqrt_n = int(n**0.5)
    for i in range(2, sqrt_n + 1):
        if n % i == 0:
            total += i
            if i != n // i:
                total += n // i
    return total

amicables = set()
for a in range(2, 10000):
    b = sum_of_divisors(a)
    if b != a and sum_of_divisors(b) == a:
        amicables.add(a)
        amicables.add(b)

print(sum(amicables))
```

## Initial approach is clean, simple, and efficient enough for the given input size.
