## Initial approach

* relatively prime to n? it's key to understand what this is.
* gcd(a,n) = 1 --> a is relatively prime to n
* lets define gdc with a function

```python
import math
```

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a
```

* now define euler's totient function by using gdc.

```python
def totient(n):
    count = 0
    for i in range(1, n + 1):
        if gcd(i, n) == 1:
            count += 1
    return count
```

## Optimized:

```python
result_n = 1
prime = 2

while result_n * prime <= 1000000:
    result_n *= prime
    prime += 1
    while True:
        for i in range(2, int(math.sqrt(prime)) + 1):
            if prime % i == 0:
                break
        else:
            break
        prime += 1

print("Result:", result_n)
```
