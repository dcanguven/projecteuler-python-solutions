## Initial approach

* For each number n, calculate n.
* Convert the cube to a string and sort its digits.
* Group cube numbers with the same sorted digits.
* If a group has exactly 5 cube numbers, stop and return the smallest one.

```python
for n in range(1, 11):
    cube = n ** 3
    print(f"{n} = {cube}")
```

```python
for n in range(1, 11):
    cube = n ** 3
    sorted_digits = ''.join(sorted(str(cube)))
    print(f"{n} = {cube}, sorted digits: {sorted_digits}")
```

```python
cube_groups = dict()

for n in range(1, 10000):
    cube = n ** 3
    key = ''.join(sorted(str(cube)))
    if key in cube_groups:
        cube_groups[key].append(cube)
    else:
        cube_groups[key] = [cube]

for key, group in cube_groups.items():
    if len(group) > 1:
        print(f"{key}: {group}")
```

```python
cube_groups = dict()
n = 1

while True:
    cube = n ** 3
    key = ''.join(sorted(str(cube)))
    if key in cube_groups:
        cube_groups[key].append(cube)
        if len(cube_groups[key]) == 5:
            print("Result:", min(cube_groups[key]))
            break
    else:
        cube_groups[key] = [cube]
    n += 1
```
