## Initial approach

* each ball touches the inner wall, so its center sits at a fixed offset from the pipe axis
* two neighbouring balls should sit on opposite sides of the pipe to reduce the needed spacing
* once you use opposite sides, the needed spacing between two neighbours depends only on the sum of their radii
* to make all neighbour sums as small as possible, avoid putting two big radii next to each other
* a good way is to place all even radii from large to small, then all odd radii from small to large
* add the two end radii, add all neighbour spacings, then convert to micrometres and round

```python
%%time
import math

def gap(a, b):
    return math.sqrt(200 * (a + b - 50))

def total_length_mm(order):
    s = order[0] + order[-1]
    for x, y in zip(order, order[1:]):
        s += gap(x, y)
    return s

evens_desc = list(range(50, 29, -2))
odds_asc = list(range(31, 50, 2))
order = evens_desc + odds_asc

length_mm = total_length_mm(order)
result = int(length_mm * 1000 + 0.5)

print("Result:", result)
```
