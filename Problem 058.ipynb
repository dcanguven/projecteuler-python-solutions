## Initial approach

The problem looks similar to problem 28

Spiral structre is the same.  
In problem 28, we calculate the sum of the numbers on the diagonals of the spiral.  
Now we check how many of those diagonal numbers are prime and stop iteration when the ratio falls below 10%.   

Lets start with is_prime function

```python
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5)+1, 2):
        if n % i == 0:
            return False
    return True
```

generate a loop:

* add new layer
* for each layer side length increases by 2 (1 from each side)
* for each layer, calculate the corner values
* if corner is prime, count it
* add 4 to the total diagonal count (1 for each corners)
* if the ratio of primes on diagonals is less then 10%, stop.

```python
%%time

prime_count = 0
total_diagonal = 1
side_length = 1

while True:
    side_length += 2
    for i in range(0, 4):
        corner = side_length**2 - i*(side_length - 1)
        if is_prime(corner):
            prime_count += 1
    total_diagonal += 4
    if prime_count / total_diagonal < 0.10:
        print(side_length)
        break
```
