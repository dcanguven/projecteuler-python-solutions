## Initial approach

* coresilience is C(n) = (n  (n)) / (n  1)
* for odd composite n, (n) is even, and C(n) is always bigger than 1/2, so the only possible unit fraction value is 1/2
* setting C(n) = 1/2 gives a clean condition, 2(n  (n)) = n  1, which simplifies to 2(n) = n + 1
* divide both sides by n to get (n)/n = 1/2 + 1/(2n), so the prime-factor product (1  1/p) must be just above 1/2
* if n has primes 3, 5, and 7 together then (1  1/p) already drops below 1/2, so only primes 3 and 5 can appear
* check n = 3^a and n = 3^a5^b in the equation 2(n) = n + 1, and it forces a = 1 and b = 1
* the only composite solution below 210^11 is n = 15, so the required sum is 15

```python
def solve():
    limit = 2 * 10**11
    sols = []
    n = 15
    if n < limit:
        sols.append(n)
    return sum(sols)
```

```python
%%time
result = solve()
print("Result:", result)
```
