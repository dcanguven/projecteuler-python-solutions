## Initial approach

This problem is very similar to Problem 64. In both problems:
* We generate continued fraction terms.
* We process terms to get results.
* We use similar loops and math with fractions.

Difference:
* Problem 64 counts how many.
* Problem 65 calculates a big numbers digit sum.

```python
def get_e_continued_fraction_terms(limit):
    terms = [2]
    k = 1
    while len(terms) < limit:
        terms += [1, 2 * k, 1]
        k += 1
    return terms[:limit]

def compute_numerator(limit):
    terms = get_e_continued_fraction_terms(limit)
    numerator = 1
    denominator = terms[-1]
    for term in reversed(terms[:-1]):
        numerator, denominator = denominator, term * denominator + numerator
    return denominator

numerator = compute_numerator(100)
digit_sum = sum(int(digit) for digit in str(numerator))
print(digit_sum)
```
