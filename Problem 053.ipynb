For each pair (n, r), calculate the combination value:
   - Use math.comb(n, r) if you have Python 3.8 or newer.
   - Or use the formula: n! / (r! * (n - r)!).

```python
%%time

from math import comb

count = 0
for n in range(1, 101):
    for r in range(0, n + 1):
        if comb(n, r) > 1_000_000:
            count += 1

print(count)
```
