## Initial approach

* It can be understood that the computational power is way too high for a traditional iteration.
* as we need only 10 digits, we can use pow function with mod 10**10
* first try to find the example

```python
def self_powers_traditional(n):
    total = 0
    for i in range(1,n+1):
        total += i**i
    return (total)
```

```python
self_powers_traditional(10)
```

Now try 1000 but now use pow function

```python
def self_powers_pow(n):
    total = 0
    for i in range(1,n+1):
        total+= pow(i, i, 10**10)
        total %= 10**10 # mod again
    return total
```

```python
%%time

self_powers_pow(1000)
```
