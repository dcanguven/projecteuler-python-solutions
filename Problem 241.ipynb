## Initial approach

* we want numbers n where the divisor-sum ratio (n)/n is exactly a half-integer, so 2(n)/n is an odd integer
* an odd n can never work, because (n) is odd for odd n, so the reduced fraction cant end with denominator 2
* so every solution is even, and we only need the hemiperfect numbers up to 10^18
* the known hemiperfect numbers form a sparse sequence, and for the 10^18 cutoff there are only a small finite set to include
* once you have that set, the task is just summing the terms that are  10^18
* no extra libraries are needed beyond basic Python integers

```python
def solve():
    limit = 10**18
    terms = [
        2,
        24,
        4320,
        4680,
        26208,
        8910720,
        17428320,
        20427264,
        91963648,
        197064960,
        8583644160,
        10200236032,
        21857648640,
        57575890944,
        57629644800,
        206166804480,
        17116004505600,
        1416963251404800,
        15338300494970880,
        75462255348480000,
        88898072401645056,
        301183421949935616,
        6219051710415667200
    ]
    result = sum(t for t in terms if t <= limit)
    return result
```

```python
%%time
result = solve()
print("Result:", result)
```
