## Initial approach

* a lamina uses T tiles with T equals 4k times m where k is border thickness and m is outer minus thickness
* valid shapes satisfy m at least k plus 1 so the inner hole is at least 1 by 1
* for each k from 1 upward add 1 to count[T] for all multiples T equals 4k times m with m from k plus 1 while T at most N
* type of T is the number of k,m pairs it has hence count[T]
* we need how many T up to one million have type between 1 and 10
* iterate and tally efficiently since total inner steps scale like N times harmonic of sqrt N
* return the tally as the answer

```python
def solve(N=10**6):
    cnt = [0] * (N + 1)
    k = 1
    while 4 * k * (k + 1) <= N:
        m = k + 1
        while True:
            t = 4 * k * m
            if t > N:
                break
            cnt[t] += 1
            m += 1
        k += 1
    result = sum(1 for t in range(1, N + 1) if 1 <= cnt[t] <= 10)
    return result
```

```python
%%time
result = solve()
print("Result:", result)
```
